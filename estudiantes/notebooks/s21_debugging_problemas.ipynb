## ğŸ› Â¿QuÃ© es Debugging?

**Debugging** es el proceso de encontrar y corregir errores (bugs) en el cÃ³digo. Todo programador pasa mÃ¡s tiempo debuggeando que escribiendo cÃ³digo nuevo.

### Los 3 tipos principales de errores:
1. **Errores de sintaxis** â†’ Python no entiende tu cÃ³digo
2. **Errores en tiempo de ejecuciÃ³n** â†’ El cÃ³digo se ejecuta pero falla
3. **Errores lÃ³gicos** â†’ El cÃ³digo funciona pero no hace lo esperado


### 1. Errores de Sintaxis

Python no puede ejecutar cÃ³digo con errores de sintaxis. Son como errores gramaticales en un idioma.


```python
# Ejemplos de errores de sintaxis comunes

# 1. Olvidar cerrar comillas
print("Hola mundo)  # Error: EOL while scanning string literal

# 2. Olvidar cerrar parÃ©ntesis
print("Hola"  # Error: unexpected EOF while parsing

# 3. IndentaciÃ³n incorrecta
def mi_funcion():
print("Hola")  # Error: expected an indented block

# 4. Usar = en vez de == en condiciones
if x = 5:  # Error: invalid syntax
    print("x es 5")
```

### 2. Errores en Tiempo de EjecuciÃ³n (Runtime Errors)

El cÃ³digo tiene sintaxis correcta, pero falla cuando se ejecuta.


```python
# Ejemplos de errores en tiempo de ejecuciÃ³n

# 1. NameError - Variable no definida
print(variable_inexistente)  # NameError: name 'variable_inexistente' is not defined

# 2. TypeError - OperaciÃ³n con tipos incorrectos
numero = 10
texto = "20"
resultado = numero + texto  # TypeError: unsupported operand type(s) for +: 'int' and 'str'

# 3. ValueError - Valor incorrecto para una funciÃ³n
numero = int("no es un nÃºmero")  # ValueError: invalid literal for int()

# 4. IndexError - Ãndice fuera de rango
lista = [1, 2, 3]
print(lista[10])  # IndexError: list index out of range

# 5. ZeroDivisionError - DivisiÃ³n por cero
resultado = 10 / 0  # ZeroDivisionError: division by zero

# 6. KeyError - Clave no existe en diccionario
diccionario = {"a": 1, "b": 2}
print(diccionario["c"])  # KeyError: 'c'
```

### 3. Errores LÃ³gicos

El cÃ³digo se ejecuta sin errores, pero no produce el resultado esperado. Son los mÃ¡s difÃ­ciles de encontrar.


```python
# Ejemplo de error lÃ³gico
# Programa para calcular el promedio de 3 nÃºmeros

def calcular_promedio(a, b, c):
    # Error lÃ³gico: estÃ¡ sumando pero no dividiendo
    return a + b + c  # Â¡Falta dividir entre 3!

# El cÃ³digo funciona, pero da resultado incorrecto
print(f"Promedio de 10, 20, 30: {calcular_promedio(10, 20, 30)}")
print("Resultado esperado: 20")
print("Resultado obtenido: 60 (incorrecto)")
```

## ğŸ”§ TÃ©cnicas de Debugging en Colab

### 1. Print Debugging (La tÃ©cnica mÃ¡s simple)
Insertar `print()` para ver el valor de variables en diferentes puntos del cÃ³digo.


```python
# Ejemplo: Debugging con print()

def calcular_descuento(precio, tiene_descuento):
    print(f"DEBUG: precio={precio}, tiene_descuento={tiene_descuento}")

    if tiene_descuento:
        precio_final = precio * 0.9
    else:
        precio_final = precio

    print(f"DEBUG: precio_final={precio_final}")
    return precio_final

# Test
resultado = calcular_descuento(1000, True)
print(f"Resultado final: {resultado}")
```

### 2. Try-Except para capturar errores

Usar `try-except` para manejar errores y obtener mÃ¡s informaciÃ³n.


```python
# Ejemplo: Debugging con try-except

def dividir_numeros(a, b):
    try:
        resultado = a / b
        return resultado
    except ZeroDivisionError as e:
        print(f"ERROR: DivisiÃ³n por cero - a={a}, b={b}")
        return None
    except TypeError as e:
        print(f"ERROR: Tipo incorrecto - a={type(a)}, b={type(b)}")
        return None

# Tests
print("Test 1 (normal):", dividir_numeros(10, 2))
print("Test 2 (divisiÃ³n por cero):", dividir_numeros(10, 0))
print("Test 3 (tipo incorrecto):", dividir_numeros(10, "2"))
```

### 3. Assertions para validar suposiciones

Usar `assert` para verificar que tu cÃ³digo hace lo que esperas.


```python
# Ejemplo: Debugging con assertions

def aplicar_descuento(precio, porcentaje):
    # Verificar precondiciones
    assert precio > 0, f"Precio debe ser positivo, recibido: {precio}"
    assert 0 <= porcentaje <= 100, f"Porcentaje debe estar entre 0 y 100, recibido: {porcentaje}"

    descuento = precio * (porcentaje / 100)
    precio_final = precio - descuento

    # Verificar postcondiciÃ³n
    assert precio_final >= 0, f"Precio final negativo: {precio_final}"

    return precio_final

# Tests
print("Descuento normal:", aplicar_descuento(1000, 20))

# Esto generarÃ¡ un AssertionError
# print("Descuento invÃ¡lido:", aplicar_descuento(-100, 20))
```

## ğŸ¬ POO: Errores Comunes ğŸ†•

Al trabajar con clases, hay errores especÃ­ficos que encontrarÃ¡s frecuentemente.

### Error 1: Falta `self` en `__init__`

```python
class Estudiante:
    def __init__(self, nombre):
        self.nombre = nombre

# Funciona correctamente
e = Estudiante("Ana")
print(e.nombre)
```

### Error 2: Olvidar `self.` al acceder atributos

```python
class Estudiante:
    def mostrar(self):
        print(f"Nombre: {self.nombre}")  # Â¡Usa self.nombre!

e = Estudiante("Ana")
e.nombre = "Ana"
e.mostrar()
```

### Error 3: Confundir clase con objeto

```python
class Producto:
    def __init__(self, nombre):
        self.nombre = nombre

# Â¡Crea un objeto primero!
p = Producto("Laptop")
print(p.nombre)  # âœ… Usar p, no Producto
```

### Error 4: Modificar objeto por accidente

```python
p1 = Producto("Laptop", 1000)
p2 = Producto("Mouse", 25)

# p1 sigue siendo Laptop
print(f"p1: {p1.nombre}")
print(f"p2: {p2.nombre}")
```

### 4. DepuraciÃ³n Paso a Paso

En Colab, puedes ejecutar celdas una por una para ver quÃ© pasa en cada paso.


```python
# Ejemplo: DepuraciÃ³n paso a paso
# Ejecuta cada bloque de cÃ³digo por separado

# Paso 1: Definir variables
nombre = "Ana"
edad = 25
puntajes = [85, 90, 78, 92]

print("Paso 1 completado")
print(f"Variables: nombre={nombre}, edad={edad}, puntajes={puntajes}")

# Paso 2: Calcular promedio
suma = 0
for puntaje in puntajes:
    suma += puntaje

print(f"Paso 2 completado")
print(f"Suma: {suma}")

# Paso 3: Calcular promedio
promedio = suma / len(puntajes)

print(f"Paso 3 completado")
print(f"Promedio: {promedio}")

# Paso 4: Determinar categorÃ­a
if promedio >= 90:
    categoria = "Excelente"
elif promedio >= 80:
    categoria = "Bueno"
elif promedio >= 70:
    categoria = "Aceptable"
else:
    categoria = "Necesita mejorar"

print(f"Paso 4 completado")
print(f"CategorÃ­a: {categoria}")
```

## ğŸ§  Taller Guiado: Debugging en AcciÃ³n

Vamos a debuggear un programa juntos. El programa deberÃ­a calcular estadÃ­sticas de ventas, pero tiene varios errores.


```python
# Programa con errores - Vamos a debuggearlo juntos

ventas = [15000, 22000, 18000, 0, 31000, 28000]

def calcular_estadisticas(ventas):
    # Calcular total
    total = 0
    for venta in ventas:
        total = venta  # ERROR: deberÃ­a ser total += venta

    # Calcular promedio
    promedio = total / len(ventas)  # ERROR POTENCIAL: divisiÃ³n por cero si lista vacÃ­a

    # Encontrar venta mÃ¡xima
    max_venta = 0
    for venta in ventas:
        if venta > max_venta:
            max_venta = venta

    # Encontrar dÃ­as con ventas bajas
    dias_bajos = []
    for i in range(len(ventas)):
        if ventas[i] < 10000:  # ERROR: Ã­ndice incorrecto
            dias_bajos.append(i)

    return total, promedio, max_venta, dias_bajos

# Ejecutar funciÃ³n
resultados = calcular_estadisticas(ventas)
print(f"Total: {resultados[0]}")
print(f"Promedio: {resultados[1]}")
print(f"Venta mÃ¡xima: {resultados[2]}")
print(f"DÃ­as con ventas bajas: {resultados[3]}")
```

### Nivel C: Encontrar errores lÃ³gicos


```python
# Ejercicio C1: Encuentra el error lÃ³gico

def es_bisiesto(anio):
    # Un aÃ±o es bisiesto si:
    # - Es divisible por 4
    # - No es divisible por 100, a menos que tambiÃ©n sea divisible por 400

    if anio % 4 == 0:
        if anio % 100 == 0:
            if anio % 400 == 0:
                return True
            else:
                return False
        else:
            return True
    else:
        return True  # ERROR LÃ“GICO: deberÃ­a ser False

# Tests
print("2000 (bisiesto):", es_bisiesto(2000))  # DeberÃ­a ser True
print("1900 (no bisiesto):", es_bisiesto(1900))  # DeberÃ­a ser False
print("2024 (bisiesto):", es_bisiesto(2024))  # DeberÃ­a ser True
print("2023 (no bisiesto):", es_bisiesto(2023))  # DeberÃ­a ser False
```
