## üíæ Paso 3: Manejo de Archivos

Tu proyecto debe usar al menos:
- **Un archivo CSV** para datos principales
- **Un archivo TXT** para logs o reportes

### Patr√≥n recomendado:
1. **Al iniciar:** Cargar datos desde CSV si existe
2. **Durante ejecuci√≥n:** Trabajar con datos en memoria
3. **Al salir/guardar:** Escribir datos actualizados a CSV

````python
# Ejemplo: Cargar gastos desde CSV
import csv

def cargar_gastos_desde_csv(archivo_csv):
    '''Carga gastos desde un archivo CSV.'''
    gastos = []
    try:
        with open(archivo_csv, 'r', newline='', encoding='utf-8') as archivo:
            lector = csv.DictReader(archivo)
            for fila in lector:
                gastos.append(fila)
        print(f"‚úÖ Cargados {len(gastos)} gastos desde {archivo_csv}")
    except FileNotFoundError:
        print(f"üìÑ Archivo {archivo_csv} no encontrado. Se crear√° uno nuevo.")
    return gastos

# Ejemplo de uso
gastos = cargar_gastos_desde_csv("gastos.csv")
print(f"Total de gastos en memoria: {len(gastos)}")
````

````python
# Ejemplo: Guardar gastos a CSV
import csv

def guardar_gastos_a_csv(gastos, archivo_csv):
    '''Guarda gastos a un archivo CSV.'''
    if not gastos:
        print("‚ö†Ô∏è  No hay gastos para guardar.")
        return
    
    # Definir nombres de columnas
    campos = ["fecha", "descripcion", "monto", "categoria"]
    
    try:
        with open(archivo_csv, 'w', newline='', encoding='utf-8') as archivo:
            escritor = csv.DictWriter(archivo, fieldnames=campos)
            escritor.writeheader()
            escritor.writerows(gastos)
        print(f"üíæ Gastos guardados en {archivo_csv}")
    except Exception as e:
        print(f"‚ùå Error al guardar: {e}")

# Ejemplo de uso
gastos_ejemplo = [
    {"fecha": "2024-01-15", "descripcion": "Supermercado", "monto": "85000", "categoria": "Alimentaci√≥n"},
    {"fecha": "2024-01-16", "descripcion": "Transporte", "monto": "15000", "categoria": "Transporte"}
]

guardar_gastos_a_csv(gastos_ejemplo, "gastos_ejemplo.csv")
````

Vamos a construir juntos los primeros componentes de un Gestor de Gastos (Opci√≥n A). Puedes adaptar estos conceptos a tu proyecto.

````python
# C√ìDIGO BASE - Gestor de Gastos
# Puedes usar este c√≥digo como punto de partida

import csv
import os

# Constantes
ARCHIVO_GASTOS = "gastos.csv"
CATEGORIAS = ["Alimentaci√≥n", "Transporte", "Entretenimiento", "Servicios", "Otros"]

def cargar_gastos():
    '''Carga gastos desde archivo CSV.'''
    gastos = []
    if os.path.exists(ARCHIVO_GASTOS):
        with open(ARCHIVO_GASTOS, 'r', newline='', encoding='utf-8') as archivo:
            lector = csv.DictReader(archivo)
            for fila in lector:
                gastos.append(fila)
        print(f"‚úÖ Cargados {len(gastos)} gastos")
    else:
        print("üìÑ No hay archivo de gastos. Se crear√° uno nuevo.")
    return gastos

def mostrar_gastos(gastos):
    '''Muestra todos los gastos registrados.'''
    if not gastos:
        print("üì≠ No hay gastos registrados.")
        return
    
    print("\nüìã GASTOS REGISTRADOS:")
    print("-" * 60)
    total = 0
    for i, gasto in enumerate(gastos, 1):
        print(f"{i}. {gasto['fecha']} - {gasto['descripcion']} - ${gasto['monto']} - {gasto['categoria']}")
        total += float(gasto['monto'])
    print("-" * 60)
    print(f"üí∞ TOTAL: ${total:.0f}")

# Ejemplo de uso
gastos = cargar_gastos()
mostrar_gastos(gastos)
````
