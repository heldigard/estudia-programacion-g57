### Procesar datos de un CSV

````python
import csv

# Calcular el total de ventas
total_ventas = 0
total_productos = 0

with open('ventas.csv', 'r', encoding='utf-8') as archivo:
    lector = csv.DictReader(archivo)
    
    for venta in lector:
        cantidad = int(venta['cantidad'])
        precio = int(venta['precio'])
        subtotal = cantidad * precio
        
        total_ventas += subtotal
        total_productos += cantidad

print('=== RESUMEN DE VENTAS ===')
print(f'Total de productos vendidos: {total_productos}')
print(f'Total de ingresos: ${total_ventas:,}')
````

````python
# Encontrar ventas por vendedor
import csv

ventas_por_vendedor = {}

with open('ventas.csv', 'r', encoding='utf-8') as archivo:
    lector = csv.DictReader(archivo)
    
    for venta in lector:
        vendedor = venta['vendedor']
        subtotal = int(venta['cantidad']) * int(venta['precio'])
        
        if vendedor in ventas_por_vendedor:
            ventas_por_vendedor[vendedor] += subtotal
        else:
            ventas_por_vendedor[vendedor] = subtotal

print('=== VENTAS POR VENDEDOR ===')
for vendedor, total in ventas_por_vendedor.items():
    print(f'{vendedor}: ${total:,}')
````

## üé¨ POO: Validaciones en `__init__` üÜï

Al leer/escribir archivos, quieres asegurarte de que los datos sean **v√°lidos**.

Las clases son perfectas para esto: puedes **validar en `__init__`** y rechazar datos incorrectos.

### ¬øPor qu√© Validar?

- **Defender tus datos:** No aceptar informaci√≥n incorrecta
- **Prevenir errores:** Mejor fallar r√°pido que tener datos corruptos
- **Mensajes claros:** Decirle al usuario qu√© est√° mal

````python
# === EJEMPLO: Clase Estudiante con Validaciones ===

class Estudiante:
    def __init__(self, nombre, edad, nota):
        # Validar nombre
        if not nombre or not nombre.strip():
            raise ValueError("El nombre no puede estar vac√≠o")
        if len(nombre) < 3:
            raise ValueError("El nombre debe tener al menos 3 caracteres")
        
        # Validar edad
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un n√∫mero entero")
        if edad < 0 or edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120")
        
        # Validar nota
        if not isinstance(nota, (int, float)):
            raise TypeError("La nota debe ser un n√∫mero")
        if nota < 0 or nota > 100:
            raise ValueError("La nota debe estar entre 0 y 100")
        
        # Si todo est√° v√°lido, guardar los atributos
        self.nombre = nombre.strip()
        self.edad = edad
        self.nota = nota

# Probar la clase
try:
    estudiante1 = Estudiante("Ana", 20, 85)
    print(f"‚úÖ Estudiante v√°lido: {estudiante1.nombre}, Nota: {estudiante1.nota}")
except (ValueError, TypeError) as e:
    print(f"‚ùå Error: {e}")
````

````python
# Probar con datos inv√°lidos
print("=== Probando validaciones ===")

# Caso 1: Nombre muy corto
try:
    e = Estudiante("Jo", 20, 85)  # Nombre muy corto
except (ValueError, TypeError) as e:
    print(f"Nombre 'Jo': ‚ùå {e}")

# Caso 2: Edad inv√°lida
try:
    e = Estudiante("Mar√≠a", 150, 85)  # Edad imposible
except (ValueError, TypeError) as e:
    print(f"Edad 150: ‚ùå {e}")

# Caso 3: Nota fuera de rango
try:
    e = Estudiante("Carlos", 22, 150)  # Nota > 100
except (ValueError, TypeError) as e:
    print(f"Nota 150: ‚ùå {e}")

# Caso 4: Datos v√°lidos
try:
    e = Estudiante("Pedro", 25, 90)
    print(f"Pedro (25, 90): ‚úÖ V√°lido")
except (ValueError, TypeError) as e:
    print(f"‚ùå {e}")
````

### Validaciones al Leer de Archivos

Cuando lees datos de un archivo, usa clases para validar autom√°ticamente.

````python
import csv

class Producto:
    def __init__(self, nombre, precio, stock):
        # Validar nombre
        if not nombre or len(nombre.strip()) == 0:
            raise ValueError(f"Nombre inv√°lido: '{nombre}'")
        
        # Validar precio
        try:
            precio_num = float(precio)
            if precio_num <= 0:
                raise ValueError(f"Precio debe ser positivo: {precio}")
        except (ValueError, TypeError):
            raise TypeError(f"Precio debe ser n√∫mero: {precio}")
        
        # Validar stock
        try:
            stock_num = int(stock)
            if stock_num < 0:
                raise ValueError(f"Stock no puede ser negativo: {stock}")
        except (ValueError, TypeError):
            raise TypeError(f"Stock debe ser entero: {stock}")
        
        self.nombre = nombre.strip()
        self.precio = precio_num
        self.stock = stock_num
        
    def valor_total(self):
        """Calcula el valor total del inventario."""
        return self.precio * self.stock

# Leer productos desde CSV con validaci√≥n
def leer_productos_con_validacion(archivo_csv):
    """Lee productos y valida cada uno."""
    productos_validos = []
    errores = []
    
    with open(archivo_csv, 'r', encoding='utf-8') as archivo:
        lector = csv.DictReader(archivo)
        for i, fila in enumerate(lector, 1):
            try:
                producto = Producto(
                    fila['producto'],
                    fila['precio'],
                    fila.get('stock', 0)  # stock puede no existir
                )
                productos_validos.append(producto)
                print(f"Fila {i}: ‚úÖ {producto.nombre}")
            except (ValueError, TypeError) as e:
                errores.append((i, fila, e))
                print(f"Fila {i}: ‚ùå {e}")
    
    return productos_validos, errores

# Crear un CSV de prueba
with open('productos_test.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['producto', 'precio', 'stock'])
    writer.writerow(['Laptop', '1000000', '5'])
    writer.writerow(['Mouse', '25000', '50'])
    writer.writerow(['', '15000', '10'])  # Nombre vac√≠o
    writer.writerow(['Teclado', '-500', '30'])  # Precio negativo
    writer.writerow(['Monitor', 'abc', '3'])  # Precio no num√©rico

print("=== Leyendo productos con validaci√≥n ===")
productos, errores = leer_productos_con_validacion('productos_test.csv')
print(f"\nProductos v√°lidos: {len(productos)}")
print(f"Errores encontrados: {len(errores)}")
````

### Validaciones al Escribir en Archivos

Usa m√©todos que validan antes de guardar.

````python
class GestorContactos:
    """Gestiona contactos con validaciones."""
    
    def __init__(self, archivo_salida):
        self.archivo_salida = archivo_salida
        self.contactos = []
    
    def agregar_contacto(self, nombre, email, edad):
        """Agrega un contacto con validaciones."""
        # Validar nombre
        if not nombre or len(nombre.strip()) < 3:
            raise ValueError("Nombre debe tener al menos 3 caracteres")
        
        # Validar email (b√°sico)
        if '@' not in email:
            raise ValueError(f"Email inv√°lido: {email}")
        
        # Validar edad
        try:
            edad_num = int(edad)
            if edad_num < 0 or edad_num > 120:
                raise ValueError("Edad debe estar entre 0 y 120")
        except ValueError:
            raise TypeError("Edad debe ser un n√∫mero")
        
        contacto = {
            'nombre': nombre.strip(),
            'email': email.strip(),
            'edad': edad_num
        }
        self.contactos.append(contacto)
        return contacto
    
    def guardar_archivo(self):
        """Guarda los contactos v√°lidos en CSV."""
        with open(self.archivo_salida, 'w', newline='', encoding='utf-8') as f:
            campos = ['nombre', 'email', 'edad']
            writer = csv.DictWriter(f, fieldnames=campos)
            writer.writeheader()
            writer.writerows(self.contactos)
        print(f"‚úÖ {len(self.contactos)} contactos guardados en {self.archivo_salida}")

# Usar el gestor con validaciones
print("=== Gestor de Contactos con Validaciones ===")
gestor = GestorContactos('contactos_validados.csv')

# Agregar contactos v√°lidos
gestor.agregar_contacto('Ana Garc√≠a', 'ana@email.com', 25)
gestor.agregar_contacto('Carlos L√≥pez', 'carlos@email.com', 30)

# Intentar agregar inv√°lidos
casos_prueba = [
    ('Jo', 'jo@email.com', 20),  # Nombre muy corto
    ('Mar√≠a', 'sin-arroba', 25),  # Email sin @
    ('Pedro', 'pedro@email.com', 150),  # Edad inv√°lida
]

for nombre, email, edad in casos_prueba:
    try:
        gestor.agregar_contacto(nombre, email, edad)
    except (ValueError, TypeError) as e:
        print(f"‚ùå {nombre}: {e}")

# Guardar solo los v√°lidos
gestor.guardar_archivo()

# Verificar
print("\n=== Archivo creado ===")
with open('contactos_validados.csv', 'r', encoding='utf-8') as f:
    print(f.read())
````

## Actividad Extra (10-15 min): Resumen rapido de ventas

Objetivo: leer CSV y calcular un total de ventas.

Pasos sugeridos:
1. Abre ventas.csv con DictReader
2. Multiplica cantidad * precio
3. Suma el total general

````python
# === MINI-PROYECTO: RESUMEN DE VENTAS ===
import csv

ruta = "../datasets/ventas.csv"
total = 0

with open(ruta, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        total += int(row["precio"]) * int(row["cantidad"])

print("Total ventas:", total)
````

````python
# === ANALIZADOR DE CONTACTOS ===
import csv

def cargar_contactos(archivo_csv):
    """Carga los contactos desde un archivo CSV."""
    contactos = []
    with open(archivo_csv, 'r', encoding='utf-8') as archivo:
        lector = csv.DictReader(archivo)
        for fila in lector:
            contactos.append(fila)
    return contactos

def contar_por_ciudad(contactos):
    """Cuenta contactos por ciudad."""
    ciudades = {}
    for contacto in contactos:
        ciudad = contacto['ciudad']
        ciudades[ciudad] = ciudades.get(ciudad, 0) + 1
    return ciudades

def promedio_edad(contactos):
    """Calcula la edad promedio."""
    edades = [int(c['edad']) for c in contactos]
    return sum(edades) / len(edades)

def buscar_por_ciudad(contactos, ciudad):
    """Filtra contactos por ciudad."""
    return [c for c in contactos if c['ciudad'].lower() == ciudad.lower()]

def generar_reporte(contactos):
    """Genera un reporte completo de los contactos."""
    print('=' * 50)
    print('        REPORTE DE CONTACTOS')
    print('=' * 50)
    print(f'\nüìä Total de contactos: {len(contactos)}')
    print(f'üìÖ Edad promedio: {promedio_edad(contactos):.1f} a√±os')
    
    print('\nüèôÔ∏è Contactos por ciudad:')
    ciudades = contar_por_ciudad(contactos)
    for ciudad, cantidad in sorted(ciudades.items()):
        print(f'   {ciudad}: {cantidad}')
    
    # Encontrar el m√°s joven y el m√°s viejo
    mas_joven = min(contactos, key=lambda c: int(c['edad']))
    mas_viejo = max(contactos, key=lambda c: int(c['edad']))
    
    print(f"\nüë∂ M√°s joven: {mas_joven['nombre']} ({mas_joven['edad']} a√±os)")
    print(f"üë¥ Mayor: {mas_viejo['nombre']} ({mas_viejo['edad']} a√±os)")
    print('=' * 50)

# Ejecutar
contactos = cargar_contactos('contactos.csv')
generar_reporte(contactos)
````

````python
# Buscar contactos de Bogot√°
print('\nüîç Contactos en Bogot√°:')
bogotanos = buscar_por_ciudad(contactos, 'Bogot√°')
for c in bogotanos:
    print(f"   {c['nombre']} - {c['email']}")
````

````python
# ‚ùå INCORRECTO - El archivo queda abierto
# archivo = open('datos.txt', 'w')
# archivo.write('Hola')
# # Falta archivo.close()!

# ‚úÖ CORRECTO - Usar with
with open('datos.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('Hola')
# Se cierra autom√°ticamente
print('‚úÖ Archivo cerrado correctamente')
````

## Seccion Web (S15) - Archivos y fetch/JSON

### üåê Archivos en Python vs JavaScript

**IMPORTANTE:** JavaScript en el navegador NO puede acceder directamente
a archivos del sistema por seguridad. En su lugar, usa:

1. **fetch()** - Para cargar recursos desde la web (URLs)
2. **localStorage** - Para almacenar datos localmente en el navegador
3. **JSON** - Formato de datos universal (m√°s com√∫n que CSV en web)

---

### üìñ Lectura de Archivos

| Operaci√≥n | Python | JavaScript |
|------------|--------|------------|
| Abrir archivo | `open('datos.txt', 'r')` | `fetch('datos.txt')` |
| Leer contenido | `archivo.read()` | `response.text()` |
| Leer JSON | `json.load(archivo)` | `response.json()` |
| Cerrar archivo | `archivo.close()` (o `with`) | Autom√°tico |

**Python:**
```python
# Leer archivo de texto
with open('datos.txt', 'r', encoding='utf-8') as archivo:
    contenido = archivo.read()
print(contenido)

# Leer JSON
import json
with open('datos.json', 'r') as f:
    datos = json.load(f)
print(datos['nombre'])
```

**JavaScript:**
```javascript
// Cargar recurso web con fetch
async function cargarArchivo() {
  const response = await fetch('datos.txt');
  const contenido = await response.text();
  console.log(contenido);
}

// Cargar JSON con fetch
async function cargarJSON() {
  const response = await fetch('datos.json');
  const datos = await response.json();
  console.log(datos.nombre);
}
```

---

### üìù Escritura de Archivos

| Operaci√≥n | Python | JavaScript |
|------------|--------|------------|
| Crear archivo | `open('datos.txt', 'w')` | `localStorage.setItem()` |
| Escribir texto | `archivo.write('texto')` | No disponible (seguridad) |
| Escribir JSON | `json.dump(datos, archivo)` | `localStorage.setItem('key', JSON.stringify(datos))` |

**Python:**
```python
# Escribir archivo de texto
with open('salida.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('Hola, mundo!')

# Escribir JSON
import json
datos = {"nombre": "Ana", "edad": 25}
with open('persona.json', 'w') as f:
    json.dump(datos, f, indent=2)
```

**JavaScript:**
```javascript
// Guardar en localStorage (persiste en el navegador)
localStorage.setItem('mensaje', 'Hola, mundo!');

// Guardar JSON en localStorage
const datos = { nombre: "Ana", edad: 25 };
localStorage.setItem('persona', JSON.stringify(datos));

// Leer desde localStorage
const mensaje = localStorage.getItem('mensaje');
const persona = JSON.parse(localStorage.getItem('persona'));
```

### üîë Diferencias Clave

| Concepto | Python | JavaScript (Navegador) |
|-----------|--------|------------------------|
| Acceso a archivos | Directo al sistema | Limitado por seguridad |
| Operaciones | S√≠ncronas (bloquean) | As√≠ncronas (promesas) |
| Formato com√∫n | CSV | JSON |
| Codificaci√≥n | `encoding='utf-8'` | Siempre UTF-8 |
| Cerrar archivo | Manual o `with` | Autom√°tico |

---

### ‚ö° Async/Await en JavaScript

Python ejecuta c√≥digo secuencialmente. JavaScript usa **async/await**
para operaciones que toman tiempo (como cargar archivos):

```python
# Python - C√≥digo secuencial
datos = cargar_archivo()  # Espera aqu√≠
procesar(datos)  # Luego ejecuta
```

```javascript
// JavaScript - As√≠ncrono
async function main() {
  const datos = await cargarArchivos();  // Espera sin bloquear
  procesar(datos);  // Luego ejecuta
}
```

---

### üì° API Fetch

**fetch()** es la forma moderna de hacer peticiones HTTP en JavaScript:

```javascript
// GET - Obtener datos
const response = await fetch('https://api.example.com/datos');
const datos = await response.json();

// POST - Enviar datos
await fetch('https://api.example.com/guardar', {
  method: 'POST',
  headers: { 'Content-Type': 'application/json' },
  body: JSON.stringify({ nombre: 'Ana', edad: 25 })
});
```

Equivalente en Python (usando requests):
```python
import requests

# GET
response = requests.get('https://api.example.com/datos')
datos = response.json()

# POST
requests.post('https://api.example.com/guardar', json={
    'nombre': 'Ana', 'edad': 25
})
```
