### 1. Diccionarios (`dict`)

Un **diccionario** es una colecciÃ³n de pares **clave-valor**. A diferencia de las listas que usan Ã­ndices numÃ©ricos (0, 1, 2...), los diccionarios usan **claves** que pueden ser strings o nÃºmeros.

---

#### ğŸ“– AnalogÃ­a: Un diccionario real

Imagina un diccionario de espaÃ±ol:
- Buscas la palabra **"gato"** (clave) â†’ encuentras su definiciÃ³n **"Animal felino..."** (valor)
- Buscas la palabra **"casa"** (clave) â†’ encuentras su definiciÃ³n **"Edificio para habitar"** (valor)

En Python es lo mismo:
```python
persona = {
    "nombre": "Ana",        # clave: "nombre", valor: "Ana"
    "edad": 25,             # clave: "edad", valor: 25
    "ciudad": "Madrid"      # clave: "ciudad", valor: "Madrid"
}
```

---

#### ğŸ”‘ Claves importantes:

1. **Las claves son Ãºnicas**: No puede haber dos claves iguales
2. **Los valores pueden repetirse**: Varios diccionarios pueden tener el mismo valor
3. **Acceso rÃ¡pido**: Es mÃ¡s rÃ¡pido buscar por clave que recorrer una lista
4. **Orden**: Desde Python 3.7, los diccionarios mantienen el orden de inserciÃ³n

````python
# === EJEMPLO 1: Crear y acceder a un diccionario ===

print("ğŸ“– EJEMPLO 1: Crear un diccionario")
print("-" * 40)

# Crear un diccionario de una persona
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Madrid",
    "hobbies": ["lectura", "viajes"]
}

print("Diccionario completo:")
print(persona)
print(f"Tipo: {type(persona)}")
print()

# === ACCEDER A VALORES ===
print("ğŸ”‘ ACCEDER A VALORES POR CLAVE")
print("-" * 40)

# Usar corchetes [] con la clave
print(f"persona['nombre']  â†’  {persona['nombre']}")
print(f"persona['edad']    â†’  {persona['edad']}")
print(f"persona['ciudad']  â†’  {persona['ciudad']}")
print(f"persona['hobbies'] â†’  {persona['hobbies']}")
print()

# === MODIFICAR VALORES ===
print("âœï¸  MODIFICAR UN VALOR")
print("-" * 40)

print(f"Edad original: {persona['edad']}")
persona["edad"] = 26  # Cambiar el valor
print(f"Edad nueva: {persona['edad']}")
print()

# === AGREGAR NUEVAS CLAVES ===
print("â• AGREGAR UNA NUEVA CLAVE")
print("-" * 40)

print(f"Diccionario antes: {list(persona.keys())}")
persona["email"] = "ana@email.com"  # Agregar nueva clave-valor
print(f"Diccionario despuÃ©s: {list(persona.keys())}")
print(f"Email agregado: {persona['email']}")
````

### MÃ©todos Ãºtiles de diccionarios

````python
# === EJEMPLO 2: MÃ©todos Ãºtiles de diccionarios ===

persona = {"nombre": "Carlos", "edad": 30, "ciudad": "BogotÃ¡"}

print("ğŸ“– DICCIONARIO ORIGINAL")
print("-" * 40)
print(persona)
print()

# === 1. keys() - Obtener todas las claves ===
print("ğŸ”‘ 1. keys() - Todas las claves")
print("-" * 40)
claves = list(persona.keys())
print(f"CÃ³digo: list(persona.keys())")
print(f"Resultado: {claves}")
print()

# === 2. values() - Obtener todos los valores ===
print("ğŸ’ 2. values() - Todos los valores")
print("-" * 40)
valores = list(persona.values())
print(f"CÃ³digo: list(persona.values())")
print(f"Resultado: {valores}")
print()

# === 3. items() - Obtener pares clave-valor ===
print("ğŸ”— 3. items() - Pares clave-valor")
print("-" * 40)
pares = list(persona.items())
print(f"CÃ³digo: list(persona.items())")
print(f"Resultado: {pares}")
print()

# === 4. get() - Acceso seguro (no da error si no existe) ===
print("ğŸ›¡ï¸  4. get() - Acceso seguro")
print("-" * 40)
print("CÃ³digo: persona.get('email')")
print(f"Resultado: {persona.get('email')}")  # None (no da error)
print()
print("CÃ³digo: persona.get('nombre')")
print(f"Resultado: {persona.get('nombre')}")  # Carlos
print()

# === 5. in - Verificar si una clave existe ===
print("â“ 5. in - Â¿Existe la clave?")
print("-" * 40)
print(f"CÃ³digo: 'nombre' in persona")
print(f"Resultado: {'nombre' in persona}")  # True
print()
print(f"CÃ³digo: 'email' in persona")
print(f"Resultado: {'email' in persona}")  # False
print()

# === 6. pop() - Eliminar y retornar valor ===
print("ğŸ—‘ï¸  6. pop() - Eliminar una clave")
print("-" * 40)
print(f"Antes: {persona}")
eliminado = persona.pop("ciudad")
print(f"CÃ³digo: persona.pop('ciudad')")
print(f"Valor eliminado: {eliminado}")
print(f"DespuÃ©s: {persona}")
````

### 2. Sets (Conjuntos) - Listas SIN duplicados ğŸ†•

Un **set** (conjunto) es una colecciÃ³n **sin orden** que **NO permite elementos duplicados**.

---

#### ğŸ“– AnalogÃ­a: Una bolsa de canicas

Imagina que tienes una bolsa con canicas de colores:
- Metes: roja, azul, roja, verde, azul, roja
- En la bolsa: roja, azul, verde (Â¡las duplicadas desaparecen!)

En Python:
```python
canicas = {"roja", "azul", "roja", "verde", "azul", "roja"}
# Resultado: {"roja", "azul", "verde"}
```

---

#### ğŸ¯ Â¿CuÃ¡ndo usar sets?

1. **Eliminar duplicados** de una lista
2. **Verificar pertenencia** (Â¿existe este elemento?) - mÃ¡s rÃ¡pido que en listas
3. **Operaciones matemÃ¡ticas**: uniÃ³n, intersecciÃ³n, diferencia

---

#### âš¡ Ventaja importante: BÃºsqueda ultra rÃ¡pida

```python
# Lista - busca uno por uno (lento con muchos datos)
lista = [1, 2, 3, ..., 1000000]
if 999999 in lista:  # Recorre hasta encontrarlo

# Set - acceso directo (instantÃ¡neo)
conjunto = {1, 2, 3, ..., 1000000}
if 999999 in conjunto:  # Â¡Encuentra instantÃ¡neamente!
```

````python
# === EJEMPLO 1: Eliminar duplicados con set ===

print("ğŸ¯ PROBLEMA: Tenemos una lista con duplicados")
print("-" * 40)

# Lista original con duplicados
numeros_con_repetidos = [1, 2, 3, 2, 1, 4, 5, 3, 6, 5]

print(f"Lista original: {numeros_con_repetidos}")
print(f"TamaÃ±o: {len(numeros_con_repetidos)} elementos")
print()

print("âœ¨ SOLUCIÃ“N: Convertir a set")
print("-" * 40)

# Convertir a set (los duplicados se eliminan automÃ¡ticamente)
numeros_unicos = set(numeros_con_repetidos)

print(f"CÃ³digo: set(lista)")
print(f"Resultado: {numeros_unicos}")
print(f"TamaÃ±o: {len(numeros_unicos)} elementos")
print()

print("ğŸ”„ Convertir de vuelta a lista (opcional)")
print("-" * 40)

lista_sin_duplicados = list(numeros_unicos)
print(f"CÃ³digo: list(set(lista))")
print(f"Resultado: {lista_sin_duplicados}")
print()

# === EJEMPLO CON STRINGS ===
print("ğŸ“ EJEMPLO CON NOMBRES")
print("-" * 40)

nombres = ["Ana", "Carlos", "Ana", "MarÃ­a", "Carlos", "Pedro", "Ana"]
print(f"Lista con duplicados: {nombres}")

nombres_unicos = list(set(nombres))
print(f"Solo nombres Ãºnicos: {nombres_unicos}")
print(f"Antes: {len(nombres)} elementos")
print(f"DespuÃ©s: {len(nombres_unicos)} elementos")
````

### 3. Tuplas (`tuple`) - Listas Inmutables ğŸ†•

Una **tupla** es como una lista, pero **NO se puede modificar** despuÃ©s de crearla. Es **inmutable** (inmutable = que no cambia).

---

#### ğŸ“– AnalogÃ­a: Una cÃ¡psula del tiempo

Una **lista** es como un cuaderno mÃ¡gico:
- âœ… Puedes borrar lo que escribiste
- âœ… Puedes agregar mÃ¡s pÃ¡ginas
- âœ… Puedes cambiar el contenido

Una **tupla** es como una cÃ¡psula del tiempo:
- âŒ No puedes cambiar lo que adentro pusiste
- âŒ No puedes agregar ni quitar elementos
- âœ… Pero el contenido estarÃ¡ siempre seguro

---

#### ğŸ¯ Â¿Por quÃ© querrÃ­as algo inmutable?

1. **Proteger datos importantes**: Coordenadas GPS, fechas, constantes
2. **Claves de diccionario**: Las listas NO pueden ser claves, las tuplas SÃ
3. **Retornar mÃºltiples valores**: `return x, y` devuelve una tupla
4. **Performance**: Las tuplas usan menos memoria que las listas

---

#### ğŸ“Œ Ejemplo real: Coordenadas GPS

```python
# Estas coordenadas NO deberÃ­an cambiar nunca
ubicacion_bogota = (4.6097, -74.0817)

# âŒ Esto darÃ­a error (protecciÃ³n)
# ubicacion_bogota[0] = 5.0  # Error!

# âœ… Crear una nueva tupla si necesitas cambiar
nueva_ubicacion = (5.0, -74.0817)
```

````python
# === EJEMPLO 1: Crear y acceder a tuplas ===

print("ğŸ”’ CREAR TUPLAS")
print("=" * 50)

# Diferentes formas de crear tuplas
coordenadas = (4.5, -3.2)           # Tupla de coordenadas GPS
colores_rgb = (255, 128, 0)        # Tupla de color RGB
dia_semana = ("lunes", "martes", "miÃ©rcoles", "jueves", "viernes")  # Tupla de dÃ­as

print("Coordenadas GPS:", coordenadas)
print("Color RGB:", colores_rgb)
print("DÃ­as de la semana:", dia_semana)
print(f"Tipo: {type(coordenadas)}")
print()

# === ACCEDER POR ÃNDICE (igual que las listas) ===
print("ğŸ“ ACCEDER POR ÃNDICE")
print("-" * 40)

print(f"coordenadas[0]  â†’  {coordenadas[0]} (latitud)")
print(f"coordenadas[1]  â†’  {coordenadas[1]} (longitud)")
print(f"dia_semana[-1]  â†’  {dia_semana[-1]} (Ãºltimo dÃ­a)")
print()

# === DESEMPAQUETADO (UNPACKING) - Muy Ãºtil! ===
print("ğŸ DESEMPAQUETADO DE TUPLAS")
print("-" * 40)
print("Permite asignar cada elemento de la tupla a variables separadas")

# Desempaquetar coordenadas
x, y = coordenadas
print(f"CÃ³digo: x, y = {coordenadas}")
print(f"Resultado: x = {x}, y = {y}")
print()

# Desempaquetar color RGB
print("CÃ³digo: rojo, verde, azul = (255, 128, 0)")
rojo, verde, azul = colores_rgb
print(f"Resultado: R={rojo}, G={verde}, B={azul}")
print()

# === CASO DE USO REAL: Retornar mÃºltiples valores ===
print("ğŸ“Š CASO PRÃCTICO: FunciÃ³n que retorna mÃºltiples valores")
print("-" * 40)

def obtener_estadisticas(lista):
    """Retorna mÃºltiples valores como una tupla"""
    minimo = min(lista)
    maximo = max(lista)
    promedio = sum(lista) / len(lista)
    return minimo, maximo, promedio  # Python lo empaqueta como tupla

notas = [85, 92, 78, 95, 88]
resultado = obtener_estadisticas(notas)

print(f"Notas: {notas}")
print(f"Resultado: {resultado}")
print(f"Tipo: {type(resultado)}")
print()

# Desempaquetar el resultado
nota_min, nota_max, nota_prom = obtener_estadisticas(notas)
print(f"Nota mÃ­nima: {nota_min}")
print(f"Nota mÃ¡xima: {nota_max}")
print(f"Nota promedio: {nota_prom:.1f}")
````

### Inmutabilidad: Listas vs Tuplas

**Ejercicio B1:** Sistema de inventario con diccionarios

Crea un sistema simple de inventario que:
1. Tenga un diccionario con productos y cantidades
2. Permita agregar stock a un producto
3. Permita vender (reducir) stock de un producto
4. Muestre un reporte del inventario

````python
# === EJERCICIO B1: Sistema de Inventario con Diccionarios ===

# Inventario inicial (producto: cantidad)
inventario = {
    "laptop": 5,
    "mouse": 20,
    "teclado": 15,
    "monitor": 8
}

print("ğŸ“¦ INVENTARIO INICIAL")
print("=" * 40)
for producto, cantidad in inventario.items():
    print(f"{producto}: {cantidad} unidades")
print()

# === 1. AGREGAR STOCK ===
def agregar_stock(inventario, producto, cantidad):
    """Agrega stock a un producto existente o lo crea si no existe."""
    if producto in inventario:
        inventario[producto] += cantidad
        print(f"âœ… Agregadas {cantidad} unidades de '{producto}'")
    else:
        inventario[producto] = cantidad
        print(f"âœ… Producto '{producto}' creado con {cantidad} unidades")

# Agregar stock
agregar_stock(inventario, "mouse", 10)  # Producto existe
agregar_stock(inventario, "impresora", 3)  # Nuevo producto
print()

# === 2. VENDER (REDUCIR) STOCK ===
def vender(inventario, producto, cantidad):
    """Reduce el stock de un producto si hay suficiente."""
    if producto not in inventario:
        print(f"âŒ Error: El producto '{producto}' no existe")
        return False
    
    if inventario[producto] >= cantidad:
        inventario[producto] -= cantidad
        print(f"âœ… Vendidas {cantidad} unidades de '{producto}'")
        return True
    else:
        print(f"âŒ Error: Stock insuficiente de '{producto}' (hay {inventario[producto]})")
        return False

# Vender productos
vender(inventario, "laptop", 2)  # Venta exitosa
vender(inventario, "monitor", 10)  # Stock insuficiente
vender(inventario, "cable", 1)  # Producto no existe
print()

# === 3. REPORTE DEL INVENTARIO ===
print("ğŸ“Š REPORTE ACTUALIZADO")
print("=" * 40)
print(f"Total de productos diferentes: {len(inventario)}")
print(f"Total de unidades: {sum(inventario.values())}")
print()

for producto, cantidad in sorted(inventario.items()):
    estado = "ğŸŸ¢" if cantidad > 10 else "ğŸŸ¡" if cantidad > 5 else "ğŸ”´"
    print(f"{estado} {producto}: {cantidad} unidades")
````

## ğŸ¬ POO: Validaciones en `__init__` ğŸ†•

Al leer/escribir archivos, quieres asegurarte de que los datos sean **vÃ¡lidos**.

Las clases son perfectas para esto: puedes **validar en `__init__`** y rechazar datos incorrectos.

### Â¿Por quÃ© Validar?

- **Defender tus datos:** No aceptar informaciÃ³n incorrecta
- **Prevenir errores:** Mejor fallar rÃ¡pido que tener datos corruptos
- **Mensajes claros:** Decirle al usuario quÃ© estÃ¡ mal

````python
# === EJEMPLO: Clase Estudiante con Validaciones ===

class Estudiante:
    def __init__(self, nombre, edad, nota):
        # Validar nombre
        if not nombre or not nombre.strip():
            raise ValueError("El nombre no puede estar vacÃ­o")
        if len(nombre) < 3:
            raise ValueError("El nombre debe tener al menos 3 caracteres")
        
        # Validar edad
        if not isinstance(edad, int):
            raise TypeError("La edad debe ser un nÃºmero entero")
        if edad < 0 or edad > 120:
            raise ValueError("La edad debe estar entre 0 y 120")
        
        # Validar nota
        if not isinstance(nota, (int, float)):
            raise TypeError("La nota debe ser un nÃºmero")
        if nota < 0 or nota > 100:
            raise ValueError("La nota debe estar entre 0 y 100")
        
        # Si todo estÃ¡ vÃ¡lido, guardar los atributos
        self.nombre = nombre.strip()
        self.edad = edad
        self.nota = nota

# Probar la clase
try:
    estudiante1 = Estudiante("Ana", 20, 85)
    print(f"âœ… Estudiante vÃ¡lido: {estudiante1.nombre}, Nota: {estudiante1.nota}")
except (ValueError, TypeError) as e:
    print(f"âŒ Error: {e}")
````

````python
# Probar con datos invÃ¡lidos
print("=== Probando validaciones ===")

# Caso 1: Nombre muy corto
try:
    e = Estudiante("Jo", 20, 85)  # Nombre muy corto
except (ValueError, TypeError) as e:
    print(f"Nombre 'Jo': âŒ {e}")

# Caso 2: Edad invÃ¡lida
try:
    e = Estudiante("MarÃ­a", 150, 85)  # Edad imposible
except (ValueError, TypeError) as e:
    print(f"Edad 150: âŒ {e}")

# Caso 3: Nota fuera de rango
try:
    e = Estudiante("Carlos", 22, 150)  # Nota > 100
except (ValueError, TypeError) as e:
    print(f"Nota 150: âŒ {e}")

# Caso 4: Datos vÃ¡lidos
try:
    e = Estudiante("Pedro", 25, 90)
    print(f"Pedro (25, 90): âœ… VÃ¡lido")
except (ValueError, TypeError) as e:
    print(f"âŒ {e}")
````

### Validaciones al Leer de Archivos

Cuando lees datos de un archivo, usa clases para validar automÃ¡ticamente.

````python
import csv

class Producto:
    def __init__(self, nombre, precio, stock):
        # Validar nombre
        if not nombre or len(nombre.strip()) == 0:
            raise ValueError(f"Nombre invÃ¡lido: '{nombre}'")
        
        # Validar precio
        try:
            precio_num = float(precio)
            if precio_num <= 0:
                raise ValueError(f"Precio debe ser positivo: {precio}")
        except (ValueError, TypeError):
            raise TypeError(f"Precio debe ser nÃºmero: {precio}")
        
        # Validar stock
        try:
            stock_num = int(stock)
            if stock_num < 0:
                raise ValueError(f"Stock no puede ser negativo: {stock}")
        except (ValueError, TypeError):
            raise TypeError(f"Stock debe ser entero: {stock}")
        
        self.nombre = nombre.strip()
        self.precio = precio_num
        self.stock = stock_num
        
    def valor_total(self):
        """Calcula el valor total del inventario."""
        return self.precio * self.stock

# Leer productos desde CSV con validaciÃ³n
def leer_productos_con_validacion(archivo_csv):
    """Lee productos y valida cada uno."""
    productos_validos = []
    errores = []
    
    with open(archivo_csv, 'r', encoding='utf-8') as archivo:
        lector = csv.DictReader(archivo)
        for i, fila in enumerate(lector, 1):
            try:
                producto = Producto(
                    fila['producto'],
                    fila['precio'],
                    fila.get('stock', 0)  # stock puede no existir
                )
                productos_validos.append(producto)
                print(f"Fila {i}: âœ… {producto.nombre}")
            except (ValueError, TypeError) as e:
                errores.append((i, fila, e))
                print(f"Fila {i}: âŒ {e}")
    
    return productos_validos, errores

# Crear un CSV de prueba
with open('productos_test.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['producto', 'precio', 'stock'])
    writer.writerow(['Laptop', '1000000', '5'])
    writer.writerow(['Mouse', '25000', '50'])
    writer.writerow(['', '15000', '10'])  # Nombre vacÃ­o
    writer.writerow(['Teclado', '-500', '30'])  # Precio negativo
    writer.writerow(['Monitor', 'abc', '3'])  # Precio no numÃ©rico

print("=== Leyendo productos con validaciÃ³n ===")
productos, errores = leer_productos_con_validacion('productos_test.csv')
print(f"\nProductos vÃ¡lidos: {len(productos)}")
print(f"Errores encontrados: {len(errores)}")
````

### Validaciones al Escribir en Archivos

Usa mÃ©todos que validan antes de guardar.

````python
class GestorContactos:
    """Gestiona contactos con validaciones."""
    
    def __init__(self, archivo_salida):
        self.archivo_salida = archivo_salida
        self.contactos = []
    
    def agregar_contacto(self, nombre, email, edad):
        """Agrega un contacto con validaciones."""
        # Validar nombre
        if not nombre or len(nombre.strip()) < 3:
            raise ValueError("Nombre debe tener al menos 3 caracteres")
        
        # Validar email (bÃ¡sico)
        if '@' not in email:
            raise ValueError(f"Email invÃ¡lido: {email}")
        
        # Validar edad
        try:
            edad_num = int(edad)
            if edad_num < 0 or edad_num > 120:
                raise ValueError("Edad debe estar entre 0 y 120")
        except ValueError:
            raise TypeError("Edad debe ser un nÃºmero")
        
        contacto = {
            'nombre': nombre.strip(),
            'email': email.strip(),
            'edad': edad_num
        }
        self.contactos.append(contacto)
        return contacto
    
    def guardar_archivo(self):
        """Guarda los contactos vÃ¡lidos en CSV."""
        with open(self.archivo_salida, 'w', newline='', encoding='utf-8') as f:
            campos = ['nombre', 'email', 'edad']
            writer = csv.DictWriter(f, fieldnames=campos)
            writer.writeheader()
            writer.writerows(self.contactos)
        print(f"âœ… {len(self.contactos)} contactos guardados en {self.archivo_salida}")

# Usar el gestor con validaciones
print("=== Gestor de Contactos con Validaciones ===")
gestor = GestorContactos('contactos_validados.csv')

# Agregar contactos vÃ¡lidos
gestor.agregar_contacto('Ana GarcÃ­a', 'ana@email.com', 25)
gestor.agregar_contacto('Carlos LÃ³pez', 'carlos@email.com', 30)

# Intentar agregar invÃ¡lidos
casos_prueba = [
    ('Jo', 'jo@email.com', 20),  # Nombre muy corto
    ('MarÃ­a', 'sin-arroba', 25),  # Email sin @
    ('Pedro', 'pedro@email.com', 150),  # Edad invÃ¡lida
]

for nombre, email, edad in casos_prueba:
    try:
        gestor.agregar_contacto(nombre, email, edad)
    except (ValueError, TypeError) as e:
        print(f"âŒ {nombre}: {e}")

# Guardar solo los vÃ¡lidos
gestor.guardar_archivo()

# Verificar
print("\n=== Archivo creado ===")
with open('contactos_validados.csv', 'r', encoding='utf-8') as f:
    print(f.read())
````

## Actividad Extra (10-15 min): Resumen rapido de ventas

Objetivo: leer CSV y calcular un total de ventas.

Pasos sugeridos:
1. Abre ventas.csv con DictReader
2. Multiplica cantidad * precio
3. Suma el total general

````python
# === MINI-PROYECTO: RESUMEN DE VENTAS ===
import csv

ruta = "../datasets/ventas.csv"
total = 0

with open(ruta, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        total += int(row["precio"]) * int(row["cantidad"])

print("Total ventas:", total)
````

````python
# === ANALIZADOR DE CONTACTOS ===
import csv

def cargar_contactos(archivo_csv):
    """Carga los contactos desde un archivo CSV."""
    contactos = []
    with open(archivo_csv, 'r', encoding='utf-8') as archivo:
        lector = csv.DictReader(archivo)
        for fila in lector:
            contactos.append(fila)
    return contactos

def contar_por_ciudad(contactos):
    """Cuenta contactos por ciudad."""
    ciudades = {}
    for contacto in contactos:
        ciudad = contacto['ciudad']
        ciudades[ciudad] = ciudades.get(ciudad, 0) + 1
    return ciudades

def promedio_edad(contactos):
    """Calcula la edad promedio."""
    edades = [int(c['edad']) for c in contactos]
    return sum(edades) / len(edades)

def buscar_por_ciudad(contactos, ciudad):
    """Filtra contactos por ciudad."""
    return [c for c in contactos if c['ciudad'].lower() == ciudad.lower()]

def generar_reporte(contactos):
    """Genera un reporte completo de los contactos."""
    print('=' * 50)
    print('        REPORTE DE CONTACTOS')
    print('=' * 50)
    print(f'\nğŸ“Š Total de contactos: {len(contactos)}')
    print(f'ğŸ“… Edad promedio: {promedio_edad(contactos):.1f} aÃ±os')
    
    print('\nğŸ™ï¸ Contactos por ciudad:')
    ciudades = contar_por_ciudad(contactos)
    for ciudad, cantidad in sorted(ciudades.items()):
        print(f'   {ciudad}: {cantidad}')
    
    # Encontrar el mÃ¡s joven y el mÃ¡s viejo
    mas_joven = min(contactos, key=lambda c: int(c['edad']))
    mas_viejo = max(contactos, key=lambda c: int(c['edad']))
    
    print(f"\nğŸ‘¶ MÃ¡s joven: {mas_joven['nombre']} ({mas_joven['edad']} aÃ±os)")
    print(f"ğŸ‘´ Mayor: {mas_viejo['nombre']} ({mas_viejo['edad']} aÃ±os)")
    print('=' * 50)

# Ejecutar
contactos = cargar_contactos('contactos.csv')
generar_reporte(contactos)
````
