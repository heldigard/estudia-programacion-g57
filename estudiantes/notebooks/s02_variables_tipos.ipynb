### Paso 2: Tipos int, float y str

### Paso 2.5: El Tipo Booleano (bool)

Los booleanos solo tienen dos valores posibles: `True` (verdadero) o `False` (falso). Son fundamentales para condicionales.

````python
# Valores booleanos
es_estudiante = True
tiene_trabajo = False

print(f"¬øEs estudiante? {es_estudiante}")
print(f"¬øTiene trabajo? {tiene_trabajo}")
print(f"Tipo: {type(es_estudiante)}")

# Comparaciones devuelven booleanos
print(f"
5 > 3: {5 > 3}")
print(f"10 == 10: {10 == 10}")
print(f"'Python' == 'python': {'Python' == 'python'}")
````

````python
# int - n√∫mero entero
edad = 25
print("Edad:", edad)
print("Tipo:", type(edad))

print()  # L√≠nea en blanco

# float - n√∫mero decimal
precio = 99.99
print("Precio:", precio)
print("Tipo:", type(precio))

print()

# str - texto (string)
ciudad = "Bogot√°"
print("Ciudad:", ciudad)
print("Tipo:", type(ciudad))
````

### Paso 3: La funci√≥n `input()` - Pedir datos al usuario

````python
# input() muestra un mensaje y espera que escribas algo
nombre = input("¬øCu√°l es tu nombre? ")
print("¬°Hola,", nombre + "!")
````

### ‚ö†Ô∏è ¬°IMPORTANTE! `input()` siempre devuelve texto

````python
# Aunque escribas un n√∫mero, input() lo guarda como texto
edad = input("¬øCu√°ntos a√±os tienes? ")
print("Escribiste:", edad)
print("El tipo es:", type(edad))  # <-- Ser√° 'str', no 'int'
````

### Paso 4: Convertir tipos de datos

````python
# Para hacer matem√°ticas, necesitas convertir el texto a n√∫mero

# int() convierte a n√∫mero entero
edad_texto = "25"
edad_numero = int(edad_texto)
print("Tipo original:", type(edad_texto))
print("Tipo convertido:", type(edad_numero))

print()

# float() convierte a n√∫mero decimal
precio_texto = "99.99"
precio_numero = float(precio_texto)
print("Precio como n√∫mero:", precio_numero)

print()

# str() convierte a texto
numero = 42
texto = str(numero)
print("Ahora es texto:", texto, "- Tipo:", type(texto))
````

### Paso 4.5: Conversi√≥n entre N√∫meros y isinstance()

````python
# Conversi√≥n entre int y float
entero = 42
decimal = float(entero)  # 42.0
print(f"int -> float: {entero} -> {decimal}")

decimal2 = 3.99
entero2 = int(decimal2)  # 3 (¬°trunca, no redondea!)
print(f"float -> int: {decimal2} -> {entero2}")

# Verificar tipo con isinstance()
print(f"
¬ø42 es int? {isinstance(42, int)}")
print(f"¬ø3.14 es float? {isinstance(3.14, float)}")
print(f"¬ø'Hola' es str? {isinstance('Hola', str)}")
print(f"¬øTrue es bool? {isinstance(True, bool)}")

# isinstance puede verificar m√∫ltiples tipos
valor = 42
print(f"
¬ø{valor} es int o float? {isinstance(valor, (int, float))}")
````

**üí° Tip:** Formatear n√∫meros grandes con separador de miles:
```python
numero = 1000000
print(f"{numero:,}")  # 1,000,000
```

````python
# Formato con separador de miles
salario = 3500000
poblacion = 5000000

print(f"Salario: ${salario:,}")
print(f"Poblaci√≥n: {poblacion:,} habitantes")

# Con decimales
pi = 3.14159
print(f"Pi redondeado: {pi:.2f}")
print(f"Salario con 2 decimales: ${salario:,.2f}")
````

### Paso 5: F-strings - La forma m√°s f√°cil de formatear texto

````python
# Los f-strings empiezan con f"..." y permiten incluir variables con {}
nombre = "Mar√≠a"
edad = 28
ciudad = "Medell√≠n"

# Sin f-string (complicado)
print("Hola, " + nombre + ". Tienes " + str(edad) + " a√±os y vives en " + ciudad + ".")

# Con f-string (¬°mucho m√°s f√°cil!)
print(f"Hola, {nombre}. Tienes {edad} a√±os y vives en {ciudad}.")
````

## ‚≠ê Actividad Estrella: Calculadora de Edad

Vamos a crear un programa que pide tu a√±o de nacimiento y calcula tu edad.

````python
# === CALCULADORA DE EDAD ===

# Pedir datos al usuario
nombre = input("¬øC√≥mo te llamas? ")
anio_nacimiento = input("¬øEn qu√© a√±o naciste? ")

# Convertir el a√±o a n√∫mero
anio_nacimiento = int(anio_nacimiento)

# Calcular edad
anio_actual = 2026
edad = anio_actual - anio_nacimiento

# Mostrar resultado con f-string
print(f"\n¬°Hola, {nombre}!")
print(f"Si naciste en {anio_nacimiento}, tienes aproximadamente {edad} a√±os.")
````
