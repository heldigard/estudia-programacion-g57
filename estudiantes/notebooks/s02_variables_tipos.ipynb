### Paso 2.5: Conversi√≥n entre int y float

Cuando conviertes entre n√∫meros, ten en cuenta:
- `int()` **trunca** (corta la parte decimal), NO redondea
- `float()` agrega `.0` a un entero

````python
# Conversi√≥n entre int y float (¬°cuidado con la truncaci√≥n!)
entero = 42
decimal = float(entero)  # 42.0
print(f"int -> float: {entero} -> {decimal}")

decimal2 = 3.99
entero2 = int(decimal2)  # 3 (¬°trunca, no redondea!)
print(f"float -> int: {decimal2} -> {entero2}")

# Verificar tipo con isinstance()
print(f"\n¬ø42 es int? {isinstance(42, int)}")
print(f"¬ø3.14 es float? {isinstance(3.14, float)}")
print(f"¬ø'Hola' es str? {isinstance('Hola', str)}")
print(f"¬øTrue es bool? {isinstance(True, bool)}")

# isinstance puede verificar m√∫ltiples tipos
valor = 42
print(f"\n¬ø{valor} es int o float? {isinstance(valor, (int, float))}")
````

````python
# === DEMO: LOS 4 TIPOS DE DATOS B√ÅSICOS ===

print("=== 1. int - N√öMERO ENTERO ===")
edad = 25
print("Edad:", edad)
print("Tipo:", type(edad))

print("\n=== 2. float - N√öMERO DECIMAL ===")
precio = 99.99
print("Precio:", precio)
print("Tipo:", type(precio))

print("\n=== 3. str - TEXTO (STRING) ===")
ciudad = "Bogot√°"
print("Ciudad:", ciudad)
print("Tipo:", type(ciudad))

print("\n=== 4. bool - VERDADERO/FALSO ===")
es_estudiante = True
tiene_trabajo = False
print("¬øEs estudiante?", es_estudiante)
print("¬øTiene trabajo?", tiene_trabajo)
print("Tipo:", type(es_estudiante))

# Comparaciones devuelven booleanos
print("\n--- Comparaciones que devuelven bool ---")
print(f"5 > 3: {5 > 3}")
print(f"10 == 10: {10 == 10}")
print(f"'Python' == 'python': {'Python' == 'python'}")
````

**üí° Tip:** Formatear n√∫meros grandes con separador de miles usando *f-strings*

- **`:`** ‚Äî separa la expresi√≥n de la especificaci√≥n de formato dentro de las llaves. Sintaxis: `variable:formato`.
- **`,`** ‚Äî en la especificaci√≥n de formato indica **separador de miles** (agrupa por miles).

Ejemplos:
```python
numero = 1000000
print(f"{numero}")    # 1000000 (sin formato)
print(f"{numero:,}")  # 1,000,000 (con separador de miles)

numero2 = 12345.678
print(f"{numero2:,.2f}")  # 12,345.68 (separa miles y muestra 2 decimales)
```

> **Nota:** Puedes combinar `,` con otros especificadores (`.2f`, `d`, etc.) para ajustar decimales u otros formatos.

````python
# Ejemplos pr√°cticos de formato (ejecuta esta celda)
numero = 1000000
print("sin formato:", f"{numero}")
print("con miles  :", f"{numero:,}")

numero2 = 12345.6789
print("float sin  :", f"{numero2}")
print("float form.:", f"{numero2:,.2f}")

peque = 42
print("peque con  :", f"{peque:,}")  # v√°lido pero no agrega separador si no hay miles
````

````python
# Formato con separador de miles
salario = 3500000
poblacion = 5000000

print(f"Salario: ${salario:,}")
print(f"Poblaci√≥n: {poblacion:,} habitantes")

# Con decimales
pi = 3.14159
print(f"Pi redondeado: {pi:.2f}")
print(f"Salario con 2 decimales: ${salario:,.2f}")
````

### Paso 5: F-strings - La forma m√°s f√°cil de formatear texto

````python
# Los f-strings empiezan con f"..." y permiten incluir variables con {}
nombre = "Mar√≠a"
edad = 28
ciudad = "Medell√≠n"

# Sin f-string (complicado)
print("Hola, " + nombre + ". Tienes " + str(edad) + " a√±os y vives en " + ciudad + ".")

# Con f-string (¬°mucho m√°s f√°cil!)
print(f"Hola, {nombre}. Tienes {edad} a√±os y vives en {ciudad}.")
````

### 1. HTML (HyperText Markup Language)

**HTML** define **qu√© hay** en la p√°gina (t√≠tulos, p√°rrafos, botones, inputs).

**Ejemplo de HTML:**
```html
<!-- Esto es un comentario: no se muestra -->
<h1>Mi P√°gina Web</h1>
<p>Este es un p√°rrafo de texto.</p>
<button>Click aqu√≠</button>
<input type="text" placeholder="Escribe algo">
```

**Conceptos clave de HTML:**

| Etiqueta | ¬øQu√© hace? | Ejemplo |
|----------|------------|---------|
| `<h1>` a `<h6>` | T√≠tulos (h1 = m√°s grande) | `<h1>T√≠tulo Principal</h1>` |
| `<p>` | P√°rrafo de texto | `<p>Hola mundo</p>` |
| `<div>` | Contenedor (caja invisible) | `<div>Contenido aqu√≠</div>` |
| `<span>` | Contenedor en l√≠nea | `<span>texto destacado</span>` |
| `<button>` | Bot√≥n clickeable | `<button>Click</button>` |
| `<input>` | Campo de entrada | `<input type="text">` |
| `<label>` | Etiqueta para inputs | `<label>Nombre:</label>` |
| `<section>` | Secci√≥n tem√°tica | `<section><h2>T√≠tulo</h2></section>` |

**Atributos importantes:**
- `id="nombre"` - Identificador √∫nico (como un DNI)
- `class="clase"` - Clase para CSS (puede repetirse)
- `placeholder="texto"` - Texto de ejemplo en inputs
- `type="text/number/email"` - Tipo de input

### 2. CSS (Cascading Style Sheets)

**CSS** define **c√≥mo se ve** la p√°gina (colores, tama√±os, posiciones).

**Ejemplo de CSS:**
```css
/* Esto es un comentario */
h1 {
  color: blue;          /* Color del texto */
  font-size: 32px;      /* Tama√±o de fuente */
  text-align: center;   /* Alineaci√≥n */
}

button {
  background: green;    /* Color de fondo */
  color: white;         /* Color del texto */
  padding: 10px 20px;   /* Espacio interno */
  border-radius: 8px;   /* Bordes redondeados */
}
```

**Conceptos clave de CSS:**

| Propiedad | ¬øQu√© hace? | Valores comunes |
|-----------|------------|-----------------|
| `color` | Color del texto | `red`, `#ff0000`, `rgb(255,0,0)` |
| `background` | Color de fondo | `blue`, `#0000ff` |
| `font-size` | Tama√±o de texto | `16px`, `1.2rem`, `120%` |
| `margin` | Espacio externo | `10px`, `20px 10px` |
| `padding` | Espacio interno | `15px`, `10px 20px` |
| `border-radius` | Bordes redondeados | `8px`, `50%` |
| `display` | Tipo de visualizaci√≥n | `block`, `inline`, `flex` |
| `text-align` | Alineaci√≥n de texto | `left`, `center`, `right` |

**Selectores CSS:**
```css
/* Por etiqueta */
button { ... }

/* Por ID (usa #) */
#miBoton { ... }

/* Por clase (usa .) */
.mi-clase { ... }

/* Combinado */
button.primary { ... }  /* botones con clase "primary" */
```

### 3. JavaScript (El lenguaje de programaci√≥n web)

**JavaScript** hace la p√°gina **interactiva** (responde a clicks, calcula, muestra mensajes).

**Ejemplo de JavaScript:**
```javascript
// Declarar variables (similar a Python)
let nombre = "Ana";          // Similar a: nombre = "Ana"
const edad = 25;             // Similar a: edad = 25 (constante)
```
