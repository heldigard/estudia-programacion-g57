## üìö Conceptos Clave

### print() vs return

| `print()` | `return` |
|-----------|----------|
| Muestra en pantalla | Devuelve un valor |
| No se puede usar el resultado | Se puede guardar y usar |
| Para mostrar informaci√≥n al usuario | Para que el programa use el resultado |

### Analog√≠a: El Panadero ü•ñ
- **print()** = El panadero te muestra el pan y dice "¬°Mira qu√© bonito!"
- **return** = El panadero te da el pan para que te lo lleves

## üé¨ Ejemplo Guiado: print vs return

### Paso 1: La diferencia fundamental

````python
# Funci√≥n que MUESTRA (print)
def sumar_y_mostrar(a, b):
    resultado = a + b
    print(f"La suma es: {resultado}")

# Funci√≥n que DEVUELVE (return)
def sumar_y_devolver(a, b):
    resultado = a + b
    return resultado

# Probemos ambas
print("--- Con print ---")
x = sumar_y_mostrar(5, 3)  # Imprime el resultado
print(f"El valor de x es: {x}")  # x es None!

print("\n--- Con return ---")
y = sumar_y_devolver(5, 3)  # No imprime nada, pero devuelve
print(f"El valor de y es: {y}")  # y tiene el valor 8!
````

### Paso 2: ¬øPor qu√© return es tan √∫til?

````python
def calcular_area(base, altura):
    """Devuelve el √°rea de un rect√°ngulo."""
    return base * altura

# Podemos usar el resultado de muchas formas:

# 1. Guardar en variable
area1 = calcular_area(5, 3)
print(f"√Årea 1: {area1}")

# 2. Usar directamente en un c√°lculo
area_total = calcular_area(5, 3) + calcular_area(4, 2)
print(f"√Årea total: {area_total}")

# 3. Usar en una condici√≥n
if calcular_area(10, 5) > 40:
    print("El √°rea es mayor a 40")

# 4. Usar en un print (formateado como queramos)
print(f"El √°rea de 7x4 es {calcular_area(7, 4)} metros cuadrados")
````

### Paso 3: Funciones matem√°ticas con return

````python
def cuadrado(n):
    """Devuelve el cuadrado de un n√∫mero."""
    return n ** 2

def cubo(n):
    """Devuelve el cubo de un n√∫mero."""
    return n ** 3

def promedio(lista):
    """Devuelve el promedio de una lista de n√∫meros."""
    return sum(lista) / len(lista)

def es_par(n):
    """Devuelve True si el n√∫mero es par, False si es impar."""
    return n % 2 == 0

# Probamos las funciones
print(f"Cuadrado de 5: {cuadrado(5)}")
print(f"Cubo de 3: {cubo(3)}")
print(f"Promedio de [10, 20, 30]: {promedio([10, 20, 30])}")
print(f"¬ø8 es par? {es_par(8)}")
print(f"¬ø7 es par? {es_par(7)}")
````

### Paso 4: return termina la funci√≥n

````python
def verificar_edad(edad):
    """Verifica si es mayor de edad y devuelve un mensaje."""
    if edad < 0:
        return "Error: edad no puede ser negativa"
    
    if edad < 18:
        return "Menor de edad"
    
    return "Mayor de edad"
    
    # Esto NUNCA se ejecuta (est√° despu√©s del return)
    print("Esta l√≠nea nunca se ejecuta")

print(verificar_edad(-5))
print(verificar_edad(15))
print(verificar_edad(25))
````

### Paso 5: Devolver m√∫ltiples valores

````python
def calcular_estadisticas(numeros):
    """Devuelve m√≠nimo, m√°ximo y promedio de una lista."""
    minimo = min(numeros)
    maximo = max(numeros)
    promedio = sum(numeros) / len(numeros)
    return minimo, maximo, promedio

# Capturar los tres valores
notas = [85, 92, 78, 90, 88]
nota_min, nota_max, nota_prom = calcular_estadisticas(notas)

print(f"Notas: {notas}")
print(f"M√≠nima: {nota_min}")
print(f"M√°xima: {nota_max}")
print(f"Promedio: {nota_prom}")
````

````python
def dividir_con_residuo(dividendo, divisor):
    """Devuelve el cociente y el residuo de una divisi√≥n."""
    cociente = dividendo // divisor
    residuo = dividendo % divisor
    return cociente, residuo

# Uso
c, r = dividir_con_residuo(17, 5)
print(f"17 √∑ 5 = {c} con residuo {r}")
````

### üè† *args y **kwargs - Par√°metros Variables (Opcional)

````python
# *args: recibe cualquier n√∫mero de argumentos posicionales
def sumar_todos(*numeros):
    """Suma todos los n√∫meros que reciba."""
    total = 0
    for n in numeros:
        total += n
    return total

print(sumar_todos(1, 2, 3))        # 6
print(sumar_todos(10, 20, 30, 40)) # 100
print(sumar_todos(5))             # 5

# **kwargs: recibe cualquier n√∫mero de argumentos con nombre
def crear_perfil(**datos):
    """Crea un perfil con los datos proporcionados."""
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

print("
=== Perfil ===")
crear_perfil(nombre="Ana", edad=25, ciudad="Madrid")

# Combinando ambos
def funcion_compleja(*numeros, **datos):
    print(f"N√∫meros: {numeros}")
    print(f"Datos: {datos}")

print("
=== Funci√≥n con ambos ===")
funcion_compleja(1, 2, 3, nombre="Test", activo=True)
````

### üè† Funciones Lambda (Opcional)

Las funciones lambda son funciones an√≥nimas (sin nombre) de una sola l√≠nea. Son √∫tiles pero no esenciales para principiantes.

````python
# Forma tradicional
def cuadrado_tradicional(x):
    return x ** 2

# Forma lambda (an√≥nima)
cuadrado_lambda = lambda x: x ** 2

print(f"Tradicional: {cuadrado_tradicional(5)}")
print(f"Lambda: {cuadrado_lambda(5)}")

# Lambda es √∫til para funciones simples
doble = lambda x: x * 2
suma = lambda x, y: x + y
es_par = lambda n: n % 2 == 0

print(f"
Doble de 5: {doble(5)}")
print(f"Suma 3 + 4: {suma(3, 4)}")
print(f"¬ø8 es par? {es_par(8)}")

# Lambda con funciones de orden superior
numeros = [1, 2, 3, 4, 5]
dobles = list(map(lambda x: x * 2, numeros))
print(f"
Dobles: {dobles}")
````

### üè† Recursividad (Opcional)

Una funci√≥n recursiva se llama a s√≠ misma. Debe tener un caso base para detenerse. Es un concepto importante pero avanzado.

````python
# Factorial con recursi√≥n
def factorial(n):
    """Calcula el factorial de n."""
    # Caso base
    if n <= 1:
        return 1
    # Caso recursivo
    return n * factorial(n - 1)

print("=== Factorial Recursivo ===")
for i in range(1, 6):
    print(f"{i}! = {factorial(i)}")

# Fibonacci con recursi√≥n
def fibonacci(n):
    """Calcula el n-√©simo n√∫mero de Fibonacci."""
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(f"
Fibonacci(7): {fibonacci(7)}")

# Contador regresivo recursivo
def cuenta_regresiva(n):
    if n > 0:
        print(n)
        cuenta_regresiva(n - 1)
    else:
        print("¬°Despegue! üöÄ")

print(f"
=== Cuenta Regresiva ===")
cuenta_regresiva(3)
````

## üé¨ Scope (Alcance de Variables)

### Variables locales vs globales

````python
# Variable GLOBAL (fuera de cualquier funci√≥n)
mensaje_global = "Soy global"

def mi_funcion():
    # Variable LOCAL (dentro de la funci√≥n)
    mensaje_local = "Soy local"
    print(f"Dentro de la funci√≥n: {mensaje_local}")
    print(f"Puedo ver la global: {mensaje_global}")

mi_funcion()

print(f"\nFuera de la funci√≥n: {mensaje_global}")
# print(mensaje_local)  # ‚ùå Error! mensaje_local no existe aqu√≠
````

````python
# Las variables locales no afectan a las globales con el mismo nombre
contador = 100  # Global

def incrementar():
    contador = 0  # Local (diferente de la global)
    contador += 1
    print(f"Contador local: {contador}")

incrementar()
print(f"Contador global: {contador}")  # Sigue siendo 100
````

### La soluci√≥n correcta: usar return

````python
def incrementar(valor):
    """Devuelve el valor incrementado en 1."""
    return valor + 1

contador = 100
print(f"Antes: {contador}")

contador = incrementar(contador)
print(f"Despu√©s: {contador}")

contador = incrementar(contador)
print(f"Despu√©s x2: {contador}")
````

### Python return ‚Üî JavaScript return

En JavaScript, el concepto de `return` es **casi id√©ntico** a Python. La principal diferencia es sint√°ctica.

| Concepto | Python | JavaScript |
|----------|--------|------------|
| Devolver valor | `return valor` | `return valor;` |
| Sin return | Devuelve `None` | Devuelve `undefined` |
| M√∫ltiples valores | `return a, b, c` | `return [a, b, c]` (array) |
| Return temprano | `if x: return` | `if (x) { return; }` |

#### Diferencias Importantes:

**1. M√∫ltiples valores de retorno:**

```python
# Python - devuelve tupla autom√°ticamente
def estadisticas(lista):
    return min(lista), max(lista), sum(lista)/len(lista)

minimo, maximo, prom = estadisticas([10, 20, 30])
```

```javascript
// JavaScript - debe devolver un array
function estadisticas(lista) {
    return [Math.min(...lista), Math.max(...lista), 
            lista.reduce((a,b)=>a+b,0)/lista.length];
}

const [minimo, maximo, prom] = estadisticas([10, 20, 30]);
// ^destructuring assignment^
```

**2. Return temprano (early return):**

Ambos lenguajes usan el mismo patr√≥n - cuando se ejecuta `return`, la funci√≥n termina inmediatamente:

```python
# Python
def verificar_edad(edad):
    if edad < 0:
        return "Error"  # ‚Üê termina aqu√≠
    if edad < 18:
        return "Menor"  # ‚Üê o aqu√≠
    return "Mayor"     # ‚Üê o aqu√≠
```

```javascript
// JavaScript
function verificarEdad(edad) {
    if (edad < 0) {
        return "Error";  // ‚Üê termina aqu√≠
    }
    if (edad < 18) {
        return "Menor";  // ‚Üê o aqu√≠
    }
    return "Mayor";      // ‚Üê o aqu√≠
}
```

**3. Lambda vs Arrow Functions:**

```python
# Python lambda
doble = lambda x: x * 2
```

```javascript
// JavaScript arrow function
const doble = x => x * 2;
```
