## ðŸ“š Conceptos Clave

### Operadores AritmÃ©ticos
| Operador | Nombre | Ejemplo | Resultado |
|----------|--------|---------|----------|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `5 - 3` | `2` |
| `*` | MultiplicaciÃ³n | `5 * 3` | `15` |
| `/` | DivisiÃ³n | `5 / 3` | `1.666...` |
| `//` | DivisiÃ³n entera | `5 // 3` | `1` |
| `%` | MÃ³dulo (residuo) | `5 % 3` | `2` |
| `**` | Potencia | `5 ** 3` | `125` |

### Operadores de ComparaciÃ³n
| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Diferente de | `5 != 3` | `True` |
| `>` | Mayor que | `5 > 3` | `True` |
| `<` | Menor que | `5 < 3` | `False` |
| `>=` | Mayor o igual | `5 >= 5` | `True` |
| `<=` | Menor o igual | `5 <= 3` | `False` |

### Operadores de Identidad

| Operador | Significado | Ejemplo |
|----------|-------------|--------|
| `is` | Â¿Es el mismo objeto? | `a is b` |
| `is not` | Â¿Es un objeto diferente? | `a is not b` |

````python
# is vs == (Â¡IMPORTANTE!)
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(f"a == b: {a == b}")  # True - tienen los mismos valores
print(f"a is b: {a is b}")  # False - son objetos diferentes
print(f"a is c: {a is c}")  # True - c es el mismo objeto que a

# None es un caso especial (siempre usa is)
x = None
print(f"
x == None: {x == None}")  # True, pero no recomendado
print(f"x is None: {x is None}")  # True - Â¡Esta es la forma correcta!
````

### Operadores de Pertenencia

| Operador | Significado | Ejemplo |
|----------|-------------|-------- |
| `in` | Â¿EstÃ¡ contenido en? | `"a" in "abc"` |
| `not in` | Â¿NO estÃ¡ contenido en? | `2 not in [1, 3, 5]` |

````python
# in con strings
texto = "Python es divertido"
print(f"Â¿Contiene 'Python'? {'Python' in texto}")
print(f"Â¿Contiene 'java'? {'java' in texto}")

# in con listas
frutas = ["manzana", "banana", "naranja"]
print(f"
Â¿Hay 'manzana'? {'manzana' in frutas}")
print(f"Â¿Hay 'uva'? {'uva' in frutas}")

# not in
print(f"
Â¿No hay 'uva'? {'uva' not in frutas}")

# Ejemplo prÃ¡ctico: verificar vocales
letra = "a"
vocales = "aeiou"
print(f"
Â¿'{letra}' es vocal? {letra in vocales}")
````

## ðŸŽ¬ Ejemplo Guiado: Operadores AritmÃ©ticos

### Paso 1: Los operadores bÃ¡sicos (+, -, *, /)

````python
a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"MultiplicaciÃ³n: {a} * {b} = {a * b}")
print(f"DivisiÃ³n: {a} / {b} = {a / b}")
````

### Paso 2: DivisiÃ³n entera (//) y mÃ³dulo (%)

Estos son muy Ãºtiles y a veces confunden al principio:

````python
# DivisiÃ³n entera: descarta los decimales
print(f"10 / 3 = {10 / 3}")    # DivisiÃ³n normal: 3.333...
print(f"10 // 3 = {10 // 3}")  # DivisiÃ³n entera: 3

print()

# MÃ³dulo: devuelve el RESIDUO de la divisiÃ³n
print(f"10 % 3 = {10 % 3}")   # 10 dividido 3 = 3 con residuo 1
print(f"15 % 5 = {15 % 5}")   # 15 dividido 5 = 3 con residuo 0
print(f"17 % 4 = {17 % 4}")   # 17 dividido 4 = 4 con residuo 1
````

**ðŸ’¡ Tip:** El mÃ³dulo `%` es muy Ãºtil para saber si un nÃºmero es par o impar:

````python
numero = 7
residuo = numero % 2

print(f"{numero} % 2 = {residuo}")
print(f"Si el residuo es 0, es par. Si es 1, es impar.")
print(f"Entonces {numero} es impar.")
````

### Paso 3: Potencias (**)

````python
# Potencias: base ** exponente
print(f"2 ** 3 = {2 ** 3}")      # 2 al cubo = 8
print(f"5 ** 2 = {5 ** 2}")      # 5 al cuadrado = 25
print(f"10 ** 0 = {10 ** 0}")    # Cualquier nÃºmero a la 0 = 1

# TambiÃ©n sirve para raÃ­ces (exponente fraccionario)
print(f"9 ** 0.5 = {9 ** 0.5}")  # RaÃ­z cuadrada de 9 = 3.0
print(f"27 ** (1/3) = {27 ** (1/3)}")  # RaÃ­z cÃºbica de 27 = 3.0
````

### Paso 4: Precedencia de Operadores (Â¿quÃ© se calcula primero?)

Python sigue reglas matemÃ¡ticas estÃ¡ndar:

````python
# Orden de precedencia (de mayor a menor):
# 1. () ParÃ©ntesis
# 2. ** Potencias
# 3. *, /, //, % MultiplicaciÃ³n, divisiÃ³n
# 4. +, - Suma, resta

# Ejemplo:
resultado = 2 + 3 * 4
print(f"2 + 3 * 4 = {resultado}")  # Primero 3*4=12, luego 2+12=14

# Con parÃ©ntesis cambia el orden:
resultado2 = (2 + 3) * 4
print(f"(2 + 3) * 4 = {resultado2}")  # Primero (2+3)=5, luego 5*4=20
````

## ðŸŽ¬ Operadores de ComparaciÃ³n

Los operadores de comparaciÃ³n devuelven `True` o `False` (valores booleanos):

````python
edad = 18

print(f"edad = {edad}")
print(f"Â¿edad == 18? {edad == 18}")   # Â¿Es igual a 18?
print(f"Â¿edad != 18? {edad != 18}")   # Â¿Es diferente de 18?
print(f"Â¿edad > 17? {edad > 17}")     # Â¿Es mayor que 17?
print(f"Â¿edad < 21? {edad < 21}")     # Â¿Es menor que 21?
print(f"Â¿edad >= 18? {edad >= 18}")   # Â¿Es mayor o igual a 18?
print(f"Â¿edad <= 17? {edad <= 17}")   # Â¿Es menor o igual a 17?
````

**âš ï¸ IMPORTANTE:** No confundas `=` (asignaciÃ³n) con `==` (comparaciÃ³n)

````python
x = 5      # AsignaciÃ³n: guardo 5 en x
print(x == 5)  # ComparaciÃ³n: Â¿x es igual a 5?
````

## ðŸŽ¬ Operadores LÃ³gicos (and, or, not)

Permiten combinar condiciones:

````python
edad = 25
tiene_licencia = True

# and: AMBAS condiciones deben ser True
puede_conducir = edad >= 18 and tiene_licencia
print(f"Â¿Puede conducir? {puede_conducir}")

# or: AL MENOS UNA condiciÃ³n debe ser True
es_fin_de_semana = False
es_feriado = True
puede_descansar = es_fin_de_semana or es_feriado
print(f"Â¿Puede descansar? {puede_descansar}")

# not: invierte el valor
esta_lloviendo = False
puede_salir = not esta_lloviendo
print(f"Â¿Puede salir? {puede_salir}")
````

### Tabla de Verdad

````python
print("=== AND (ambas deben ser True) ===")
print(f"True and True = {True and True}")
print(f"True and False = {True and False}")
print(f"False and True = {False and True}")
print(f"False and False = {False and False}")

print()

print("=== OR (al menos una True) ===")
print(f"True or True = {True or True}")
print(f"True or False = {True or False}")
print(f"False or True = {False or True}")
print(f"False or False = {False or False}")

print()

print("=== NOT (invierte) ===")
print(f"not True = {not True}")
print(f"not False = {not False}")
````

## ðŸŽ¬ Operadores de AsignaciÃ³n Aumentada

Son atajos para modificar una variable:

````python
# En lugar de escribir:
contador = 0
contador = contador + 1
print(f"contador = {contador}")

# Puedes escribir:
contador += 1  # Equivale a: contador = contador + 1
print(f"contador = {contador}")
````

````python
# Todos los operadores de asignaciÃ³n aumentada:
x = 10
print(f"x inicial = {x}")

x += 5   # x = x + 5
print(f"x += 5 â†’ x = {x}")

x -= 3   # x = x - 3
print(f"x -= 3 â†’ x = {x}")

x *= 2   # x = x * 2
print(f"x *= 2 â†’ x = {x}")

x /= 4   # x = x / 4
print(f"x /= 4 â†’ x = {x}")
````

## ðŸ”¢ Operadores JavaScript vs Python

Â¡Buena noticia! Los operadores en JavaScript son **CASI IDÃ‰NTICOS** a Python.

### Operadores AritmÃ©ticos - SON IGUALES ðŸŽ‰

| Operador | Python | JavaScript | Resultado |
|----------|--------|------------|----------|
| Suma | `5 + 3` | `5 + 3` | `8` |
| Resta | `5 - 3` | `5 - 3` | `2` |
| MultiplicaciÃ³n | `5 * 3` | `5 * 3` | `15` |
| DivisiÃ³n | `5 / 3` | `5 / 3` | `1.666...` |
| Potencia | `5 ** 2` | `5 ** 2` | `25` |
| MÃ³dulo | `5 % 3` | `5 % 3` | `2` |

**âš ï¸ Diferencia:** JavaScript no tiene divisiÃ³n entera `//`
- Python: `10 // 3` â†’ `3`
- JavaScript: `Math.floor(10 / 3)` â†’ `3`

### Operadores de ComparaciÃ³n - CASI IGUALES

| Operador | Python | JavaScript | Nota |
|----------|--------|------------|------|
| Igual | `5 == 5` | `5 === 5` | JS usa `===` |
| Diferente | `5 != 3` | `5 !== 3` | JS usa `!==` |
| Mayor que | `5 > 3` | `5 > 3` | Igual |
| Menor que | `5 < 3` | `5 < 3` | Igual |
| Mayor o igual | `5 >= 5` | `5 >= 5` | Igual |
| Menor o igual | `5 <= 3` | `5 <= 3` | Igual |

**âš ï¸ IMPORTANTE:** En JavaScript SIEMPRE usa `===` en lugar de `==`
- `==` en JS puede hacer conversiÃ³n de tipos (peligroso)
- `===` es comparaciÃ³n estricta (como `==` en Python)

### Operadores LÃ³gicos - DIFERENTES SÃMBOLOS, MISMA LÃ“GICA

| Python | JavaScript | Ejemplo |
|--------|------------|---------|
| `and` | `&&` | `true && false` â†’ `false` |
| `or` | `||` | `true || false` â†’ `true` |
| `not` | `!` | `!true` â†’ `false` |

### AsignaciÃ³n Aumentada - IGUALES

| Python | JavaScript |
|--------|------------|
| `x += 5` | `x += 5` |
| `x -= 3` | `x -= 3` |
| `x *= 2` | `x *= 2` |
| `x /= 4` | `x /= 4` |
