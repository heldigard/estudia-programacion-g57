## üìö Conceptos Clave

### ¬øQu√© son los Condicionales?
Los condicionales permiten que tu programa **tome decisiones**. Es como un sem√°foro:
- üü¢ Si la luz es verde ‚Üí avanza
- üî¥ Si la luz es roja ‚Üí detente

### Sintaxis B√°sica
```python
if condici√≥n:
    # c√≥digo que se ejecuta si la condici√≥n es True
```

**‚ö†Ô∏è IMPORTANTE:** El c√≥digo dentro del `if` debe estar **indentado** (4 espacios o 1 Tab)

## üé¨ Ejemplo Guiado: Tu Primer if

### Paso 1: El if m√°s simple

````python
edad = 18

if edad >= 18:
    print("Eres mayor de edad")
    print("Puedes votar")
````

````python
# ¬øQu√© pasa si la condici√≥n es falsa?
edad = 15

if edad >= 18:
    print("Eres mayor de edad")  # Esto NO se ejecuta

print("Este mensaje siempre aparece")  # Esto S√ç se ejecuta (est√° fuera del if)
````

### Paso 2: if-else (si no...)

````python
edad = 15

if edad >= 18:
    print("Eres mayor de edad")
    print("Puedes votar")
else:
    print("Eres menor de edad")
    print("Todav√≠a no puedes votar")
````

### Paso 3: if-elif-else (m√∫ltiples opciones)

````python
nota = 85

if nota >= 90:
    print("Excelente - A")
elif nota >= 80:
    print("Muy bien - B")
elif nota >= 70:
    print("Bien - C")
elif nota >= 60:
    print("Suficiente - D")
else:
    print("Reprobado - F")
````

**üí° Tip:** Las condiciones se eval√∫an en orden. Cuando una es `True`, las dem√°s se ignoran.

````python
# Prueba con diferentes valores
nota = int(input("Ingresa tu nota (0-100): "))

if nota >= 90:
    resultado = "Excelente - A"
elif nota >= 80:
    resultado = "Muy bien - B"
elif nota >= 70:
    resultado = "Bien - C"
elif nota >= 60:
    resultado = "Suficiente - D"
else:
    resultado = "Reprobado - F"

print(f"Tu calificaci√≥n: {resultado}")
````

### Paso 4: Combinando condiciones con and, or, not

````python
# AND: Ambas condiciones deben ser True
edad = 25
tiene_licencia = True

if edad >= 18 and tiene_licencia:
    print("Puedes conducir")
else:
    print("No puedes conducir")
````

````python
# OR: Al menos una condici√≥n debe ser True
es_estudiante = True
es_profesor = False

if es_estudiante or es_profesor:
    print("Tienes acceso a la biblioteca")
else:
    print("No tienes acceso")
````

````python
# NOT: Invierte la condici√≥n
esta_lloviendo = False

if not esta_lloviendo:
    print("¬°Buen d√≠a para salir!")
else:
    print("Mejor qu√©date en casa")
````

### Paso 5: Condicionales anidados (if dentro de if)

````python
edad = 20
tiene_entrada = True

if edad >= 18:
    print("Eres mayor de edad")
    if tiene_entrada:
        print("¬°Puedes entrar al concierto!")
    else:
        print("Necesitas comprar una entrada")
else:
    print("No puedes entrar - Solo mayores de 18")
````

### Paso 6: Operador Ternario

El operador ternario es una forma compacta de escribir un if-else en una l√≠nea.

````python
# Forma tradicional (if-else)
edad = 18
if edad >= 18:
    mensaje1 = "Es mayor"
else:
    mensaje1 = "Es menor"

# Operador ternario (m√°s compacto)
mensaje2 = "Es mayor" if edad >= 18 else "Es menor"

print(f"Con if-else: {mensaje1}")
print(f"Con ternario: {mensaje2}")

# Ejemplos pr√°cticos
nota = 85
estado = "Aprobado" if nota >= 60 else "Reprobado"
print(f"
Nota {nota}: {estado}")

hora = 14
saludo = "Buenos d√≠as" if hora < 12 else "Buenas tardes"
print(f"{saludo} (son las {hora} horas)")
````

### Paso 7: Match/Case (Python 3.10+)

El `match/case` es una alternativa m√°s legible a m√∫ltiples `if/elif` cuando comparas una variable con muchos valores.

````python
# Ejemplo: D√≠a de la semana
dia = 3

# Forma tradicional con if/elif
if dia == 1:
    nombre_dia1 = "Lunes"
elif dia == 2:
    nombre_dia1 = "Martes"
elif dia == 3:
    nombre_dia1 = "Mi√©rcoles"
else:
    nombre_dia1 = "Otro d√≠a"

# Forma moderna con match/case
match dia:
    case 1:
        nombre_dia2 = "Lunes"
    case 2:
        nombre_dia2 = "Martes"
    case 3:
        nombre_dia2 = "Mi√©rcoles"
    case _:
        nombre_dia2 = "Otro d√≠a"

print(f"if/elif: {nombre_dia1}")
print(f"match/case: {nombre_dia2}")

# Ejemplo con strings
estado = "aprobado"

match estado:
    case "aprobado":
        emoji = "‚úÖ"
    case "reprobado":
        emoji = "‚ùå"
    case _:
        emoji = "‚ùì"

print(f"
Estado {estado}: {emoji}")
````

**üí° Tip: Cortocircuito en operadores l√≥gicos**

- `or`: Devuelve el primer valor `True`, o el √∫ltimo si todos son `False`
- `and`: Devuelve el primer valor `False`, o el √∫ltimo si todos son `True`

````python
# Cortocircuito con OR
resultado_or = False or "Hola" or 42
print(f"False or 'Hola' or 42 = {resultado_or}")  # "Hola"

# Cortocircuito con AND
resultado_and = "Python" and 3 and True
print(f"\"Python\" and 3 and True = {resultado_and}")  # True

# Uso pr√°ctico: valores por defecto
nombre = ""  # String vac√≠o (falsy)
nombre_mostrar = nombre or "Desconocido"
print(f"
Nombre: {nombre_mostrar}")  # "Desconocido"

nombre2 = "Ana"
nombre_mostrar2 = nombre2 or "Desconocido"
print(f"Nombre: {nombre_mostrar2}")  # "Ana"
````

## ‚ö†Ô∏è Errores Comunes

### Error 1: Olvidar los dos puntos (:)

````python
# ‚ùå INCORRECTO
# if edad >= 18
#     print("Mayor de edad")

# ‚úÖ CORRECTO
edad = 18
if edad >= 18:
    print("Mayor de edad")
````

### Error 2: Indentaci√≥n incorrecta

````python
# ‚ùå INCORRECTO
# if edad >= 18:
# print("Mayor de edad")  # IndentationError!

# ‚úÖ CORRECTO
edad = 18
if edad >= 18:
    print("Mayor de edad")  # 4 espacios de indentaci√≥n
````

### Error 3: Usar = en lugar de ==

````python
# ‚ùå INCORRECTO
# if nombre = "Ana":  # SyntaxError!

# ‚úÖ CORRECTO
nombre = "Ana"
if nombre == "Ana":  # Usa == para comparar
    print("Hola Ana")
````

### Error 4: Orden incorrecto de elif

### Error 5: Rangos en orden incorrecto

Siempre evalua primero el rango mas alto; si no, nunca llegaras a las otras opciones.

````python
# ‚ùå INCORRECTO - El orden importa
nota = 95
if nota >= 60:
    print("Aprobado")  # Se ejecuta esto y se ignora el resto
elif nota >= 90:
    print("Excelente")  # Nunca llega aqu√≠

print()

# ‚úÖ CORRECTO - Condiciones m√°s restrictivas primero
nota = 95
if nota >= 90:
    print("Excelente")
elif nota >= 60:
    print("Aprobado")
````
