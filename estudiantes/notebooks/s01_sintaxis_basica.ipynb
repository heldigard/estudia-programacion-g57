### Paso 3: Texto (Strings)
El texto siempre va entre comillas (simples o dobles):

### Paso 3.5: Caracteres Especiales en Strings

Python usa el backslash `\` para crear caracteres especiales:

````python
# Salto de l√≠nea (
)
print("Primera l√≠nea
Segunda l√≠nea")

# Tabulaci√≥n (	)
print("Columna 1	Columna 2	Columna 3")

# Backslash literal (\)
print("Ruta: C:\Users\Usuario")

# Comillas dentro de strings
print("Ella dijo: \"Hola\"")
print("It's beautiful")
````

### Paso 3.6: Concatenaci√≥n y Repetici√≥n de Strings

````python
# Concatenaci√≥n con + (une sin espacios)
nombre = "Ana"
saludo = "Hola, " + nombre + "!"
print(saludo)  # Hola, Ana!

# print con comas (agrega espacios)
print("Hola,", nombre, "!")  # Hola, Ana !

# Repetici√≥n con *
print("-" * 20)  # --------------------
print("=" * 10 + " MEN√ö " + "=" * 10)
````

**üí° Tip:** La diferencia entre `+` y `,` en `print()`:
- `print("A" + "B")` ‚Üí `AB` (une strings)
- `print("A", "B")` ‚Üí `A B` (separa con espacio)

````python
print("Esto es un string con comillas dobles")
print('Esto es un string con comillas simples')
print("Mi nombre es Ana")
````

### Paso 4: N√∫meros y Operaciones
Python puede hacer matem√°ticas. ¬°No necesitas comillas para los n√∫meros!

````python
# Operaciones b√°sicas
print(5 + 3)   # Suma: 8
print(10 - 4)  # Resta: 6
print(6 * 7)   # Multiplicaci√≥n: 42
print(15 / 3)  # Divisi√≥n: 5.0
````

### Paso 5: Mezclando texto y n√∫meros
Puedes mostrar texto y n√∫meros juntos usando comas en `print()`:

````python
print("Mi edad es:", 25)
print("El resultado de 5 + 3 es:", 5 + 3)
print("Nombre:", "Carlos", "- Edad:", 30)
````

## ‚≠ê Actividad Estrella: Calculadora de Propinas

Vamos a crear algo √∫til: una calculadora que te dice cu√°nto dejar de propina en un restaurante.

````python
# === CALCULADORA DE PROPINAS ===

# Datos de entrada
cuenta = 1000            # Valor de la cuenta en pesos
propina_porcentaje = 10  # Porcentaje de propina

# C√°lculos
propina = cuenta * propina_porcentaje / 100
total = cuenta + propina

# Mostrar resultados
print("=== CALCULADORA DE PROPINAS ===")
print("Cuenta:", cuenta, "pesos")
print("Propina (", propina_porcentaje, "%):", propina, "pesos")
print("Total a pagar:", total, "pesos")
````

**üéâ ¬°Felicidades!** Acabas de escribir tu primer programa √∫til en Python.

## ‚ö†Ô∏è Errores Comunes (y c√≥mo arreglarlos)

### Error 1: Olvidar las comillas en un string

````python
# ‚ùå INCORRECTO - Causa error
# print(Hola mundo)

# ‚úÖ CORRECTO
print("Hola mundo")
````

### Error 2: Usar `print` sin par√©ntesis

````python
# ‚ùå INCORRECTO (esto era v√°lido en Python 2, pero NO en Python 3)
# print "Hola"

# ‚úÖ CORRECTO
print("Hola")
````

### Error 3: Unir texto con + cuando hay n√∫meros

````python
# ‚ùå INCORRECTO - No puedes "sumar" texto y n√∫mero
# print("Mi edad es " + 25)

# ‚úÖ CORRECTO - Usa comas
print("Mi edad es", 25)

# ‚úÖ TAMBI√âN CORRECTO - Convierte el n√∫mero a texto con str()
print("Mi edad es " + str(25))
````
