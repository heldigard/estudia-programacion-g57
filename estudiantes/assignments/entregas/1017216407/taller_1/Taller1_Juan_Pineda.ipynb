**Sistema de Calificaciones**

```python
calificacion = int(input("Ingresa tu calificacion 0-100:"))
while calificacion < 0 or calificacion > 100:
  calificacion = int(input("Ingresa una calificacion valida 0-100:"))
if calificacion >= 90:
  print("Excelente")
elif calificacion >= 80:
  print("Muy bueno")
elif calificacion >= 70:
  print("Bueno")
elif calificacion >= 60:
  print("Aceptable")
else:
  print("Reprobado. No te desanimes puedes hacerlo mejor la proxima vez")
```

**Calculadora de Descuentos**

```python
subtotal_pagar = 15000
descuento = 0
edad = int(input("Ingrese la edad: "))
dia = str(input("Es martes s/n? "))
while edad < 0 or edad > 120:
  edad = int(input("Ingrese una edad valida 0-120"))
while dia != "s" and dia != "n":
  dia = str(input("Ingrese s/n"))
if edad < 13:
  descuento = 0.5
elif edad > 12:
  descuento = 0.3
elif edad > 59:
  descuento = 0.4
if dia == "s":
  descuento += 0.1
total = subtotal_pagar - (subtotal_pagar * descuento)
print(f"El total a pagar es ${total}")
```

**Tabla de Multiplicar Personalizada**

```python
numero_base = int(input("Ingrese el numero base: "))
numero_inicial = int(input("Ingrese el numero inicial: "))
numero_final = int(input("Ingrese el numero final: "))
while numero_inicial < 1 or numero_final < numero_inicial:
  numero_inicial = int(input("Ingrese un numero inicial mayor a 1: "))
  numero_final = int(input("Ingrese un numero final mayor al inicial: "))
print(f"=== Tabla del {numero_base} ({numero_inicial}-{numero_final}) ===")
for i in range(numero_inicial, numero_final + 1):
  print(f"{numero_base} x {i} = {numero_base * i}")

```

**Gestor de Lista de Compras**

```python
# Lista de compras
lista_compras = []

def mostrar_menu():
    print("\n=== LISTA DE COMPRAS ===")
    print("1. Agregar producto")
    print("2. Ver lista")
    print("3. Buscar producto")
    print("4. Eliminar producto")
    print("5. EstadÃ­sticas")
    print("6. Salir")

while True:
    mostrar_menu()

    opcion = input("Elige una opciÃ³n (1-6): ").strip()

    if opcion == "1":
        producto = input("Ingresa el nombre del producto: ").strip()

        if producto:
            lista_compras.append(producto)
            print(f"'{producto}' agregado a la lista.")
        else:
            print("El nombre no puede estar vacÃ­o.")

    elif opcion == "2":
        if not lista_compras:
            print("La lista estÃ¡ vacÃ­a.")
        else:
            print("\nProductos en la lista:")
            for i, producto in enumerate(lista_compras, start=1):
                print(f"{i}. {producto}")

    elif opcion == "3":
        if not lista_compras:
            print("La lista estÃ¡ vacÃ­a.")
        else:
            buscar = input("Producto a buscar: ").strip()
            if buscar in lista_compras:
                posicion = lista_compras.index(buscar) + 1
                print(f"'{buscar}' encontrado en la posiciÃ³n {posicion}.")
            else:
                print(f"'{buscar}' no estÃ¡ en la lista.")

    elif opcion == "4":
        if not lista_compras:
            print("La lista estÃ¡ vacÃ­a.")
        else:
            eliminar = input("Producto a eliminar: ").strip()
            if eliminar in lista_compras:
                lista_compras.remove(eliminar)
                print(f"'{eliminar}' eliminado correctamente.")
            else:
                print(f"'{eliminar}' no existe en la lista.")

    elif opcion == "5":
        if not lista_compras:
            print("La lista estÃ¡ vacÃ­a.")
        else:
            total = len(lista_compras)
            producto_largo = max(lista_compras, key=len)
            orden_alfabetico = sorted(lista_compras)

            print("\nESTADÃSTICAS")
            print(f"Total de productos: {total}")
            print(f"Producto mÃ¡s largo: {producto_largo}")
            print("Productos en orden alfabÃ©tico:")
            for producto in orden_alfabetico:
                print(f"- {producto}")

    elif opcion == "6":
        print("Â¡Hasta luego!")
        break

    else:
        print("OpciÃ³n invÃ¡lida. Por favor elige un nÃºmero del 1 al 6.")
```
