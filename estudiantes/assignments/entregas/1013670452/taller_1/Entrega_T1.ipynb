Ejercicio 1: Calculadora de Edad (10 pts)

```python
# 1. Pida el nombre del usuario
nombre = input("¿Cuál es tu nombre? ")
# 2. Pida su año de nacimiento
año_nacimiento = int(int(input("¿En qué año naciste? ")))
# 3. Calcule su edad actual (2026)
edad = 2026 - año_nacimiento
#4. Determine si es mayor de edad (≥18)

if edad > 18:
    print("Eres mayor de edad. ✓")
else:
    print("No eres mayor de edad . ✗")

#5. Muestre un mensaje personalizado con toda la información

print(f"=========INFORMACIÓN DEL USUARIO=========")
print(f"¡Hola {nombre}!")
print(f"Tienes {edad} años")
print(f"¿Eres mayor de edad? {edad >= 18}")
```

Ejercicio 2: Conversor de Moneda (10 pts)


```python
# Pida una cantidad en pesos colombianos
pesos = int(input("¿Cuántos pesos colombianos quieres cambiar?"))

# Convierta a dólares (usa tasa: 1 USD = 4200 COP)
# Convierta a euros (usa tasa: 1 EUR = 4500 COP)
dolares = 4200
euros = 4500 
conversion_pesos_dolares = pesos / dolares
conversion_pesos_euros = pesos / euros

# Muestre el resultado de las conversiones
print("=========CONVERSIÓN DE MONEDA=========")
print(f"La cantidad de {pesos} pesos, equivale a: ")
print(f"{conversion_pesos_dolares:.2f} USD")
print(f"{conversion_pesos_euros:.2f} EUR")
```

Ejercicio 3: Sistema de Calificaciones (15 pts)



Valide que esté en el rango correcto
Clasifique según la escala:
90-100: "Excelente" (A)
80-89: "Muy Bueno" (B)
70-79: "Bueno" (C)
60-69: "Aceptable" (D)
0-59: "Reprobado" (F)
Indique si aprobó (≥60) o reprobó
Muestre mensaje de ánimo o felicitación según el caso

```python
# Pida una calificación numérica (0-100)
calificacion = int(input("Ingrese la calificación (0-100): "))

# Valide que esté en el rango correcto
if calificacion > 100 or calificacion < 0: 
    print("Ingrese una calificación válida")

# Muestre la calificación en letras según la escala
elif calificacion >=90: 
    print("Excelente - (A)")
elif calificacion >=80: 
    print("Muy Bueno - (B)")
elif calificacion >=70: 
    print("Bueno - (C)")
elif calificacion >=60: 
    print("Aceptable - (D)")
elif calificacion <=59: 
    print("Reprobado - (F)")

print("=========RESULTADO DE LA CALIFICACIÓN=========")
if calificacion >= 60:
    print("Felicidades, has aprobado el curso.")
else: 
    print("Lamentablemente, no has aprobado el curso. ¡Sigue intentando!")
```
