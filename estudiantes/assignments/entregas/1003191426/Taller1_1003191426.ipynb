**Ejercicio1: Calculadora de Edad**

```python
# Se reciben los datos del usuario
nombre_usuario = input("Ingresa tu nombre: ")
anio_nacimiento =  int(input("Ingresa tu año de nacimiento: "))

# Calculo de edad
edad = 2026 - anio_nacimiento

# Se mira si es mayor de edad y se muestra el resultado
print(f"¡Hola {nombre_usuario}!")
print(f"Tienes {edad} años.")
print("Eres mayor de edad!" if edad >= 18 else "No eres mayor de edad!") 
```

**Ejercicio 2: Conversor de Moneda**

```python
# Contantes
TASA_DOLAR = 4200
TASA_EURO = 4500

# Valor ingresado por el usuario
valor_pesos = int(input("Ingresa un valor en pesos: "))

# Conversión de pesos a dólares/euros
dolares = valor_pesos/TASA_DOLAR
euros = valor_pesos/TASA_EURO

# Resultado de la conversión
print(f"Cantidad en COP: ${valor_pesos}")
print(f"${valor_pesos:,.0f} COP equivalen a:\n- {dolares:,.2f} USD\n- {euros:,.2f} EUR")
```

**Ejercicio 3: Sistema de Calificaciones**

```python
# Se recibe la nota del usuario
nota = int(input("Ingresa tu nota (0-100)"))

# Calculo de estado según la nota
if nota >= 90:
    resultado = "Excelente (A) \n¡Aprobaste! \n¡Sigue así!"
elif nota >= 80:
    resultado = "Muy bueno (B) \n¡Aprobaste! \n¡Sigue así!"
elif nota >= 70:
    resultado = "Bueno (C) \n¡Aprobaste! \n¡Puedes dar más!"
elif nota>= 60:
    resultado = "Aceptable (D) \n¡Aprobaste! \n¡Más esfuerzo la próxima!"
else:
    resultado = "Reprobado (F) \n¡Suerte la próxima!"

# Se muestra el resultado
print(f"{nota} - {resultado}")
    
```

**Ejercicio 4: Calculadora de Descuentos**

```python
# Tarifa de entrada regular y descuentos aplicables
TARIFA = 15000
DESCUENTO_NINOS = 50/100
DESCUENTO_ESTUDIANTES = 30/100
DESCUENTO_ADULTOS_MAYORES = 40/100
DESCUENTO_MARTES = 10/100

# Entradas del usuario
edad = int(input("Ingresa tu edad: "))
es_martes = input("Hoy es martes? (s/n): ")

# Cálculo valor de la entrada
if edad >= 60:
    entrada = TARIFA*(1-DESCUENTO_ADULTOS_MAYORES)
    descuento = "Descuento Adulto Mayor: 40%"
elif edad >= 26:
    entrada = TARIFA
elif edad >= 13:
    entrada = TARIFA*(1-DESCUENTO_ESTUDIANTES)
    descuento = "Descuento Estudiantes: 30%"
else:
    entrada = TARIFA*(1-DESCUENTO_NINOS)    
    descuento = "Descuento Niños: 50%"

if es_martes == "s":
    entrada *= (1-DESCUENTO_MARTES)
    descuento_mensaje_martes = "Descuento por día Martes: 10%"
else:
    descuento_mensaje_martes = "Descuento por día Martes: NA"

#Muestra y desglose del precio
print(f"{"="*32}\n{"-"*14}CINE{"-"*14}\n{"="*32}")
print(f"Tarifa: ${TARIFA:,.0f} \n{descuento} \n{descuento_mensaje_martes}")
print(f"Tarifa Total con Descuento(s): ${entrada:,.0f}")
```
