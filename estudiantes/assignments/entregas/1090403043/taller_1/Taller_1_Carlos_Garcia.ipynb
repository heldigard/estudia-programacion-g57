```

````python
boleta = 15000
dia = input("hoy es martes? (si o no): ").lower()
if dia == "si":
    dia = float(0.1)
else:
    dia = float(0)
edad = -1
while edad < 0:
    edad = int(input("Ingresa tu edad: "))
    if edad < 0:
        print("Edad inv√°lida. Por favor ingresa una edad v√°lida.")
    else:
        if edad > 59:
            descuento_1 = boleta * (0.4 + dia)
            boleta_1 = boleta - descuento_1
            print(f"tu entrada tiene un costo de: ${boleta_1} \nTienes un descuento de: ${descuento_1}")
        elif edad > 25:
            descuento_2 = boleta * dia
            boleta_2 = boleta - descuento_2
            print(f"tu entrada tiene un costo de: ${boleta_2} \nTienes un descuento de: ${descuento_2}")
        elif edad > 12:
            descuento_3 = boleta * (0.3 + dia)
            boleta_3 = boleta - descuento_3
            print(f"tu entrada tiene un costo de: ${boleta_3} \nTienes un descuento de: ${descuento_3}")
        elif edad > 0:
            descuento_4 = boleta * (0.5 + dia)
            boleta_4 = boleta - descuento_4
            print(f"tu entrada tiene un costo de: ${boleta_4} \nTienes un descuento de: ${descuento_4}")
````
